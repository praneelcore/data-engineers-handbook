# Section 1: Binary Trees -- Structure and Terminology

## What is a Binary Tree?

A tree where each node has at most **2 children** (left and right).

```
        1          <- Root (level 0, depth 0)
       / \
      2   3        <- Level 1
     / \   \
    4   5   6      <- Level 2 (4, 5, 6 are leaves)
```

## Key Terms

| Term | Definition |
|------|-----------|
| **Root** | Top node (no parent) |
| **Leaf** | Node with no children |
| **Height** | Longest path from root to any leaf |
| **Depth** | Distance from root to a specific node |
| **Balanced** | Height of left and right subtrees differ by <= 1 |
| **Complete** | All levels full except possibly the last (filled left to right) |
| **Full** | Every node has 0 or 2 children |
| **BST** | Left subtree < node < right subtree (for all nodes) |

## DE Connection

- **File systems**: Directories are trees (root -> folders -> files)
- **Spark execution plans**: The query plan is a tree of operations
- **JSON/XML parsing**: Nested data is naturally tree-structured
- **B-trees**: Database indexes use balanced trees for O(log n) lookups
- **Decision trees**: ML models that split data hierarchically

````python
# Building a Binary Tree
print("BINARY TREE -- Building from Scratch")
print("=" * 60)

class TreeNode:
    """A node in a binary tree."""
    def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right

    def __repr__(self):
        return f"TreeNode({self.val})"

# Build the example tree:
#       1
#      / \
#     2   3
#    / \   \
#   4   5   6

root = TreeNode(1)
root.left = TreeNode(2)
root.right = TreeNode(3)
root.left.left = TreeNode(4)
root.left.right = TreeNode(5)
root.right.right = TreeNode(6)

print("  Tree structure:")
print("       1")
print("      / \\")
print("     2   3")
print("    / \\   \\")
print("   4   5   6")
print()
print(f"  Root: {root.val}")
print(f"  Root's left child: {root.left.val}")
print(f"  Root's right child: {root.right.val}")
print(f"  Leaf nodes: 4, 5, 6 (no children)")

# Helper to build tree from list (level-order)
def build_tree(values):
    """Build binary tree from level-order list. None = no node."""
    if not values or values[0] is None:
        return None
    from collections import deque
    root = TreeNode(values[0])
    queue = deque([root])
    i = 1
    while queue and i < len(values):
        node = queue.popleft()
        if i < len(values) and values[i] is not None:
            node.left = TreeNode(values[i])
            queue.append(node.left)
        i += 1
        if i < len(values) and values[i] is not None:
            node.right = TreeNode(values[i])
            queue.append(node.right)
        i += 1
    return root
````

# Section 2: Tree Traversals -- The Four Ways to Visit Every Node

## The Four Traversals

```
Tree:     1
         / \
        2   3
       / \
      4   5

Inorder   (Left, Root, Right): 4, 2, 5, 1, 3  <- Sorted order for BST!
Preorder  (Root, Left, Right): 1, 2, 4, 5, 3  <- Good for copying/serializing
Postorder (Left, Right, Root): 4, 5, 2, 3, 1  <- Good for deleting tree
Level-order (BFS):             1, 2, 3, 4, 5  <- Level by level
```

## When to Use Each

| Traversal | Use Case |
|-----------|----------|
| **Inorder** | Get BST elements in sorted order |
| **Preorder** | Copy/serialize tree, evaluate expressions |
| **Postorder** | Delete tree, calculate subtree values bottom-up |
| **Level-order** | Find depth, shortest path, level-by-level processing |

````python
# All four traversals
from collections import deque

print("TREE TRAVERSALS")
print("=" * 60)

# Build tree: 1, 2, 3, 4, 5
root = TreeNode(1)
root.left = TreeNode(2)
root.right = TreeNode(3)
root.left.left = TreeNode(4)
root.left.right = TreeNode(5)

# 1. Inorder (Left, Root, Right) -- recursive
def inorder(root):
    """Left -> Root -> Right. Gives sorted order for BST."""
    if not root:
        return []
    return inorder(root.left) + [root.val] + inorder(root.right)

# 2. Preorder (Root, Left, Right) -- recursive
def preorder(root):
    """Root -> Left -> Right. Good for serialization."""
    if not root:
        return []
    return [root.val] + preorder(root.left) + preorder(root.right)

# 3. Postorder (Left, Right, Root) -- recursive
def postorder(root):
    """Left -> Right -> Root. Good for deletion."""
    if not root:
        return []
    return postorder(root.left) + postorder(root.right) + [root.val]

# 4. Level-order (BFS) -- iterative with queue
def level_order(root):
    """Level by level using a queue (BFS)."""
    if not root:
        return []
    result = []
    queue = deque([root])
    while queue:
        node = queue.popleft()
        result.append(node.val)
        if node.left:
            queue.append(node.left)
        if node.right:
            queue.append(node.right)
    return result

print("  Tree:     1")
print("           / \\")
print("          2   3")
print("         / \\")
print("        4   5")
print()
print(f"  Inorder   (L, Root, R): {inorder(root)}")
print(f"  Preorder  (Root, L, R): {preorder(root)}")
print(f"  Postorder (L, R, Root): {postorder(root)}")
print(f"  Level-order (BFS):      {level_order(root)}")
````

````python
# Level-order by LEVELS (return list of lists)
print("\nLEVEL ORDER -- By Levels (list of lists)")
print("=" * 60)

def level_order_by_level(root):
    """
    Return values grouped by level.
    Time: O(n), Space: O(n)
    """
    if not root:
        return []
    result = []
    queue = deque([root])

    while queue:
        level_size = len(queue)
        level = []

        for _ in range(level_size):
            node = queue.popleft()
            level.append(node.val)
            if node.left:
                queue.append(node.left)
            if node.right:
                queue.append(node.right)

        result.append(level)

    return result

# Build a bigger tree for demo
root = build_tree([3, 9, 20, None, None, 15, 7])
result = level_order_by_level(root)
print("  Tree: [3, 9, 20, None, None, 15, 7]")
print(f"  Level order: {result}")
print("  Level 0: [3], Level 1: [9, 20], Level 2: [15, 7]")
````

# Section 3: Common Tree Problems

Most tree problems follow a recursive pattern:
1. **Base case**: What to return for `None` (empty node)?
2. **Recursive case**: Process left subtree, process right subtree, combine.

This is the **divide and conquer** approach applied to trees.

````python
# Tree Problem 1: Maximum Depth (Height)
print("TREE PROBLEM 1: Maximum Depth")
print("=" * 60)

def max_depth(root):
    """
    Find the height of the tree.
    Time: O(n), Space: O(h) where h = height (recursion stack)
    """
    if not root:
        return 0
    left_depth = max_depth(root.left)
    right_depth = max_depth(root.right)
    return 1 + max(left_depth, right_depth)

# Test
root = build_tree([3, 9, 20, None, None, 15, 7])
print(f"  Tree: [3, 9, 20, None, None, 15, 7]")
print(f"  Max depth: {max_depth(root)}")

root2 = build_tree([1, 2, 3, 4, 5])
print(f"  Tree: [1, 2, 3, 4, 5]")
print(f"  Max depth: {max_depth(root2)}")

print("\n  Pattern: depth = 1 + max(left_depth, right_depth)")
print("  Base case: None -> 0")
````

````python
# Tree Problem 2: Invert Binary Tree (Mirror)
print("\nTREE PROBLEM 2: Invert Binary Tree")
print("=" * 60)

def invert_tree(root):
    """
    Mirror the tree (swap left and right at every node).
    Time: O(n), Space: O(h)
    """
    if not root:
        return None
    # Swap children
    root.left, root.right = root.right, root.left
    # Recursively invert subtrees
    invert_tree(root.left)
    invert_tree(root.right)
    return root

root = build_tree([4, 2, 7, 1, 3, 6, 9])
print(f"  Before: {level_order(root)}")
invert_tree(root)
print(f"  After:  {level_order(root)}")
print("  (Left and right children swapped at every level)")
````

````python
# Tree Problem 3: Same Tree / Symmetric Tree
print("\nTREE PROBLEM 3: Same Tree & Symmetric Tree")
print("=" * 60)

def is_same_tree(p, q):
    """
    Check if two trees are identical.
    Time: O(n), Space: O(h)
    """
    if not p and not q:
        return True
    if not p or not q:
        return False
    return (p.val == q.val and
            is_same_tree(p.left, q.left) and
            is_same_tree(p.right, q.right))

def is_symmetric(root):
    """
    Check if tree is a mirror of itself.
    Time: O(n), Space: O(h)
    """
    def is_mirror(left, right):
        if not left and not right:
            return True
        if not left or not right:
            return False
        return (left.val == right.val and
                is_mirror(left.left, right.right) and
                is_mirror(left.right, right.left))

    if not root:
        return True
    return is_mirror(root.left, root.right)

# Test symmetric
root = build_tree([1, 2, 2, 3, 4, 4, 3])
print(f"  Tree [1,2,2,3,4,4,3]: symmetric = {is_symmetric(root)}")
root2 = build_tree([1, 2, 2, None, 3, None, 3])
print(f"  Tree [1,2,2,None,3,None,3]: symmetric = {is_symmetric(root2)}")
````

````python
# Tree Problem 4: Path Sum
print("\nTREE PROBLEM 4: Path Sum (root to leaf)")
print("=" * 60)

def has_path_sum(root, target_sum):
    """
    Check if there is a root-to-leaf path with given sum.
    Time: O(n), Space: O(h)
    """
    if not root:
        return False

    # If leaf node, check if remaining sum equals node value
    if not root.left and not root.right:
        return root.val == target_sum

    # Recurse with reduced sum
    remaining = target_sum - root.val
    return (has_path_sum(root.left, remaining) or
            has_path_sum(root.right, remaining))

root = build_tree([5, 4, 8, 11, None, 13, 4, 7, 2, None, None, None, 1])
print(f"  has_path_sum(tree, 22) = {has_path_sum(root, 22)}")
print("  Path: 5 -> 4 -> 11 -> 2 = 22")
````

# Section 4: Binary Search Tree (BST)

## What is a BST?

A binary tree where for EVERY node:
- All values in left subtree < node value
- All values in right subtree > node value

```
        8         <- For node 8: left side all < 8, right side all > 8
       / \
      3   10      <- For node 3: left(1) < 3, right(6) > 3
     / \    \
    1   6    14
       / \
      4   7
```

## Why BSTs Matter

- **Search**: O(log n) average (like binary search on sorted array)
- **Insert**: O(log n) average
- **Delete**: O(log n) average
- **Inorder traversal gives sorted order!**

## DE Connection

- Database indexes (B-trees are generalized BSTs)
- Spark's range partitioning uses BST-like structure
- Sorted data storage in key-value stores

````python
# BST Operations
print("BST OPERATIONS")
print("=" * 60)

def bst_search(root, target):
    """
    Search for value in BST.
    Time: O(log n) average, O(n) worst (unbalanced)
    """
    if not root:
        return None
    if target == root.val:
        return root
    elif target < root.val:
        return bst_search(root.left, target)
    else:
        return bst_search(root.right, target)

def bst_insert(root, val):
    """
    Insert value into BST.
    Time: O(log n) average
    """
    if not root:
        return TreeNode(val)
    if val < root.val:
        root.left = bst_insert(root.left, val)
    else:
        root.right = bst_insert(root.right, val)
    return root

# Build BST by inserting values
root = None
values = [8, 3, 10, 1, 6, 14, 4, 7]
for v in values:
    root = bst_insert(root, v)

print(f"  Inserted: {values}")
print(f"  Inorder (sorted!): {inorder(root)}")
print()

# Search
for target in [6, 13]:
    result = bst_search(root, target)
    found = result.val if result else "NOT FOUND"
    print(f"  Search for {target}: {found}")
````

````python
# Validate BST
print("\nVALIDATE BST")
print("=" * 60)

def is_valid_bst(root, min_val=float("-inf"), max_val=float("inf")):
    """
    Check if tree is a valid BST.
    Each node must be within (min_val, max_val) range.
    Time: O(n), Space: O(h)
    """
    if not root:
        return True

    if root.val <= min_val or root.val >= max_val:
        return False

    return (is_valid_bst(root.left, min_val, root.val) and
            is_valid_bst(root.right, root.val, max_val))

# Valid BST
root1 = build_tree([8, 3, 10, 1, 6, None, 14])
print(f"  Tree [8,3,10,1,6,None,14]: valid BST = {is_valid_bst(root1)}")

# Invalid BST (5 is in right subtree of 2 but > 5 which is root)
root2 = TreeNode(5)
root2.left = TreeNode(1)
root2.right = TreeNode(4)
root2.right.left = TreeNode(3)
root2.right.right = TreeNode(6)
print(f"  Tree [5,1,4,None,None,3,6]: valid BST = {is_valid_bst(root2)}")
print("  (4 is in right subtree of 5, but 4 < 5 -- violates BST property)")
````

# Section 5: Heaps (Priority Queues)

## What is a Heap?

A heap is a complete binary tree where:
- **Min-heap**: Parent <= children (root is minimum)
- **Max-heap**: Parent >= children (root is maximum)

```
Min-Heap:        Max-Heap:
    1                9
   / \              / \
  3   2            7   8
 / \              / \
5   4            5   6
```

## Why Heaps Are Useful

| Operation | Time |
|-----------|------|
| Get min/max | O(1) -- just look at root |
| Insert | O(log n) -- bubble up |
| Remove min/max | O(log n) -- bubble down |
| Build heap from array | O(n) |

## When to Use a Heap

- **Top-K elements** (most frequent, largest, smallest)
- **Merge K sorted lists**
- **Median in a stream** (two heaps)
- **Dijkstra's shortest path**
- **Task scheduling by priority**

## Python's heapq Module

Python only provides a **min-heap**. For max-heap, negate values.

```python
import heapq

# Min-heap operations
heap = []
heapq.heappush(heap, 5)     # Insert
heapq.heappush(heap, 2)
smallest = heapq.heappop(heap)  # Remove and return smallest
peek = heap[0]                    # Peek at smallest without removing

# Max-heap trick: negate values
heapq.heappush(heap, -10)   # Insert 10 as -10
largest = -heapq.heappop(heap)  # Get 10 back
```

````python
# Heap basics with heapq
import heapq

print("HEAPS WITH PYTHON heapq")
print("=" * 60)

# Basic operations
heap = []
values = [5, 3, 8, 1, 9, 2, 7]
print(f"  Inserting: {values}")

for v in values:
    heapq.heappush(heap, v)
print(f"  Heap (internal): {heap}")
print(f"  Min element (peek): {heap[0]}")

# Pop elements in sorted order
sorted_result = []
while heap:
    sorted_result.append(heapq.heappop(heap))
print(f"  Popped in order: {sorted_result}")
print("  A heap gives you elements in sorted order via repeated pop!")

# heapify: Build heap from existing list in O(n)
print("\n  heapify -- build heap from list in O(n):")
arr = [5, 3, 8, 1, 9, 2, 7]
heapq.heapify(arr)
print(f"  After heapify: {arr}")
print(f"  Min: {arr[0]}")

# nlargest / nsmallest
data = [10, 4, 6, 8, 2, 9, 1, 7, 3, 5]
print(f"\n  Data: {data}")
print(f"  3 largest:  {heapq.nlargest(3, data)}")
print(f"  3 smallest: {heapq.nsmallest(3, data)}")
````

````python
# Heap Problem 1: Kth Largest Element
print("\nHEAP PROBLEM 1: Kth Largest Element")
print("=" * 60)

def find_kth_largest(nums, k):
    """
    Find the kth largest element.
    Use a min-heap of size k: the root is the kth largest.
    Time: O(n log k), Space: O(k)
    """
    heap = []
    for num in nums:
        heapq.heappush(heap, num)
        if len(heap) > k:
            heapq.heappop(heap)  # Remove smallest
    return heap[0]  # Root = kth largest

tests = [
    ([3, 2, 1, 5, 6, 4], 2, 5),
    ([3, 2, 3, 1, 2, 4, 5, 5, 6], 4, 4),
]
for nums, k, expected in tests:
    result = find_kth_largest(nums, k)
    status = "PASS" if result == expected else "FAIL"
    print(f"  [{status}] kth_largest({nums}, k={k}) = {result}")

print("\n  Why min-heap of size k works:")
print("  - Keep only k largest elements in heap")
print("  - Smallest of those k elements = kth largest overall")
print("  - Better than sorting all (O(n log n)) when k << n")
````

````python
# Heap Problem 2: Top K Frequent Elements
print("\nHEAP PROBLEM 2: Top K Frequent Elements")
print("=" * 60)

from collections import Counter

def top_k_frequent(nums, k):
    """
    Find the k most frequent elements.
    Time: O(n log k), Space: O(n)
    """
    freq = Counter(nums)
    # Use heap to find top k by frequency
    return heapq.nlargest(k, freq.keys(), key=freq.get)

tests = [
    ([1, 1, 1, 2, 2, 3], 2, [1, 2]),
    ([1], 1, [1]),
    ([4, 4, 4, 2, 2, 3, 3, 3, 3], 2, [3, 4]),
]
for nums, k, expected in tests:
    result = top_k_frequent(nums, k)
    status = "PASS" if sorted(result) == sorted(expected) else "FAIL"
    print(f"  [{status}] top_k_frequent({nums}, k={k}) = {result}")

print("\n  DE Connection: 'Find top 100 customers by revenue' = heap problem!")
print("  Much better than sorting 100M records to get top 100.")
````

````python
# Heap Problem 3: Merge K Sorted Lists
print("\nHEAP PROBLEM 3: Merge K Sorted Lists")
print("=" * 60)

def merge_k_sorted(lists):
    """
    Merge k sorted lists into one sorted list.
    Time: O(N log k) where N = total elements, k = number of lists
    Space: O(k) for the heap
    """
    heap = []
    result = []

    # Initialize heap with first element from each list
    for i, lst in enumerate(lists):
        if lst:
            heapq.heappush(heap, (lst[0], i, 0))
            # (value, list_index, element_index)

    while heap:
        val, list_idx, elem_idx = heapq.heappop(heap)
        result.append(val)

        # Push next element from same list
        if elem_idx + 1 < len(lists[list_idx]):
            next_val = lists[list_idx][elem_idx + 1]
            heapq.heappush(heap, (next_val, list_idx, elem_idx + 1))

    return result

lists = [[1, 4, 5], [1, 3, 4], [2, 6]]
print(f"  Lists: {lists}")
print(f"  Merged: {merge_k_sorted(lists)}")
print()
print("  DE Connection: This is how Spark merges sorted partitions!")
print("  Each partition is a sorted list, heap picks the smallest across all.")
````

````python
# SOLUTION 1: Lowest Common Ancestor
print("SOLUTION 1: Lowest Common Ancestor")
print("=" * 60)

def lowest_common_ancestor(root, p, q):
    """
    If current node is p or q, return it.
    If p and q are in different subtrees, current node is LCA.
    Time: O(n), Space: O(h)
    """
    if not root or root.val == p or root.val == q:
        return root

    left = lowest_common_ancestor(root.left, p, q)
    right = lowest_common_ancestor(root.right, p, q)

    if left and right:
        return root  # p and q are in different subtrees
    return left if left else right

root = build_tree([3, 5, 1, 6, 2, 0, 8])
lca = lowest_common_ancestor(root, 5, 1)
print(f"  LCA(5, 1) = {lca.val}")  # 3

lca = lowest_common_ancestor(root, 5, 6)
print(f"  LCA(5, 6) = {lca.val if lca else None}")  # 5 (ancestor of itself)
````

````python
# SOLUTION 2: Kth Smallest in BST
print("\nSOLUTION 2: Kth Smallest in BST")
print("=" * 60)

def kth_smallest_bst(root, k):
    """
    Inorder traversal (iterative) -- stop at kth element.
    Time: O(h + k), Space: O(h)
    """
    stack = []
    current = root
    count = 0

    while current or stack:
        # Go as far left as possible
        while current:
            stack.append(current)
            current = current.left

        current = stack.pop()
        count += 1
        if count == k:
            return current.val

        current = current.right

    return -1

root = build_tree([5, 3, 6, 2, 4])
for k in [1, 2, 3, 4, 5]:
    result = kth_smallest_bst(root, k)
    print(f"  kth_smallest(k={k}) = {result}")
print("  Inorder of BST [5,3,6,2,4]: [2, 3, 4, 5, 6]")
````

````python
# SOLUTION 3: Median from Data Stream
print("\nSOLUTION 3: Median from Data Stream (Two Heaps)")
print("=" * 60)

class MedianFinder:
    """
    Two heaps approach:
    - max_heap (lower half): stores negated values for max behavior
    - min_heap (upper half): stores normal values
    Invariant: |max_heap| == |min_heap| or |max_heap| == |min_heap| + 1
    """
    def __init__(self):
        self.max_heap = []  # Lower half (negated for max-heap)
        self.min_heap = []  # Upper half

    def addNum(self, num):
        # Always add to max_heap first
        heapq.heappush(self.max_heap, -num)

        # Ensure max_heap top <= min_heap top
        if self.min_heap and (-self.max_heap[0]) > self.min_heap[0]:
            heapq.heappush(self.min_heap, -heapq.heappop(self.max_heap))

        # Balance sizes (max_heap can be at most 1 larger)
        if len(self.max_heap) > len(self.min_heap) + 1:
            heapq.heappush(self.min_heap, -heapq.heappop(self.max_heap))
        elif len(self.min_heap) > len(self.max_heap):
            heapq.heappush(self.max_heap, -heapq.heappop(self.min_heap))

    def findMedian(self):
        if len(self.max_heap) > len(self.min_heap):
            return -self.max_heap[0]
        return (-self.max_heap[0] + self.min_heap[0]) / 2

# Test
mf = MedianFinder()
stream = [2, 3, 4, 1, 5]
print(f"  Stream: {stream}")
for num in stream:
    mf.addNum(num)
    print(f"    After adding {num}: median = {mf.findMedian()}")

print("\n  Two heaps keep lower/upper halves balanced.")
print("  addNum: O(log n), findMedian: O(1)")
````

# Section 7: Summary

## Tree Traversal Cheat Sheet

| Traversal | Order | Use Case | Implementation |
|-----------|-------|----------|----------------|
| Inorder | L, Root, R | BST sorted order | Recursive or stack |
| Preorder | Root, L, R | Serialize tree | Recursive or stack |
| Postorder | L, R, Root | Delete tree, calc subtree values | Recursive or stack |
| Level-order | Level by level | Find depth, BFS | Queue |

## Heap Cheat Sheet

| Operation | Time | Python |
|-----------|------|--------|
| Build heap | O(n) | `heapq.heapify(list)` |
| Push | O(log n) | `heapq.heappush(heap, val)` |
| Pop min | O(log n) | `heapq.heappop(heap)` |
| Peek min | O(1) | `heap[0]` |
| Top k largest | O(n log k) | `heapq.nlargest(k, data)` |
| Top k smallest | O(n log k) | `heapq.nsmallest(k, data)` |
