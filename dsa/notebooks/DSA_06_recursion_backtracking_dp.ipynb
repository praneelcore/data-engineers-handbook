# Section 1: Recursion Fundamentals

## What is Recursion?

A function that **calls itself** to solve smaller versions of the same problem.

Every recursive solution has:
1. **Base case**: When to stop (prevents infinite recursion)
2. **Recursive case**: How to break the problem into smaller subproblems

## Analogy: Russian Nesting Dolls

To count dolls: open the outer doll, count 1, then count the dolls inside.
Base case: no more dolls inside = return 0.

## The Call Stack

```
factorial(4)
  -> 4 * factorial(3)
       -> 3 * factorial(2)
            -> 2 * factorial(1)
                 -> return 1  (base case!)
            <- return 2 * 1 = 2
       <- return 3 * 2 = 6
  <- return 4 * 6 = 24
```

## Key Rule: Trust the Recursion

When writing recursive code:
1. Define what the function does clearly
2. Write the base case
3. Assume the recursive call works correctly for smaller inputs
4. Use that result to solve the current problem

Do NOT try to trace through every recursive call mentally.

````python
# Recursion basics
import sys
sys.setrecursionlimit(10000)

print("RECURSION FUNDAMENTALS")
print("=" * 60)

# Example 1: Factorial
def factorial(n):
    """n! = n * (n-1) * ... * 1. Base case: 0! = 1."""
    if n <= 1:
        return 1
    return n * factorial(n - 1)

print(f"  factorial(5) = {factorial(5)}")
print(f"  factorial(10) = {factorial(10)}")

# Example 2: Sum of list
def recursive_sum(arr):
    """Sum all elements recursively."""
    if not arr:
        return 0  # Base case: empty list
    return arr[0] + recursive_sum(arr[1:])  # First + sum of rest

print(f"  recursive_sum([1,2,3,4,5]) = {recursive_sum([1, 2, 3, 4, 5])}")

# Example 3: Power function
def power(base, exp):
    """Calculate base^exp recursively."""
    if exp == 0:
        return 1
    if exp % 2 == 0:
        half = power(base, exp // 2)
        return half * half  # O(log n) -- divide exponent by 2!
    return base * power(base, exp - 1)

print(f"  power(2, 10) = {power(2, 10)}")
print(f"  power(3, 5) = {power(3, 5)}")
````

````python
# Recursion on trees (review from DSA_04)
print("\nRECURSION ON TREES")
print("=" * 60)

class TreeNode:
    def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right

# Build: [1, 2, 3, 4, 5]
root = TreeNode(1, TreeNode(2, TreeNode(4), TreeNode(5)), TreeNode(3))

# Count nodes
def count_nodes(root):
    if not root:
        return 0
    return 1 + count_nodes(root.left) + count_nodes(root.right)

# Sum all values
def sum_tree(root):
    if not root:
        return 0
    return root.val + sum_tree(root.left) + sum_tree(root.right)

print(f"  Nodes: {count_nodes(root)}")
print(f"  Sum: {sum_tree(root)}")
print("\n  Pattern: process(node) = combine(node.val, process(left), process(right))")
print("  This is the foundation for ALL tree DP problems.")
````

# Section 2: Backtracking

## What is Backtracking?

A systematic way to explore ALL possible solutions by:
1. **Choose**: Make a choice
2. **Explore**: Recursively explore what follows from that choice
3. **Unchoose**: Undo the choice (backtrack) and try the next option

## The Template

```python
def backtrack(state, choices):
    if IS_COMPLETE(state):
        result.append(state.copy())
        return

    for choice in choices:
        if IS_VALID(choice):
            MAKE_CHOICE(state, choice)      # Choose
            backtrack(state, ...)            # Explore
            UNDO_CHOICE(state, choice)       # Unchoose (backtrack)
```

## When to Use Backtracking

- Generate ALL permutations/combinations/subsets
- Solve constraint satisfaction (Sudoku, N-Queens)
- Find ALL paths that satisfy a condition
- Any problem asking "find all valid ..." or "count all ways ..."

## Time Complexity

Usually exponential: O(2^n) for subsets, O(n!) for permutations.
Acceptable only for small inputs (n <= 20-25).

````python
# Backtracking: Generate All Subsets
print("BACKTRACKING: Generate All Subsets")
print("=" * 60)

def subsets(nums):
    """
    Generate all subsets (power set).
    Time: O(2^n * n), Space: O(n) for recursion depth
    """
    result = []

    def backtrack(start, current):
        result.append(current[:])  # Add copy of current subset

        for i in range(start, len(nums)):
            current.append(nums[i])       # Choose
            backtrack(i + 1, current)     # Explore
            current.pop()                  # Unchoose

    backtrack(0, [])
    return result

nums = [1, 2, 3]
result = subsets(nums)
print(f"  Subsets of {nums}:")
for s in result:
    print(f"    {s}")
print(f"  Total: {len(result)} subsets (2^{len(nums)} = {2**len(nums)})")
````

````python
# Backtracking: Generate All Permutations
print("\nBACKTRACKING: Generate All Permutations")
print("=" * 60)

def permutations(nums):
    """
    Generate all permutations.
    Time: O(n! * n), Space: O(n)
    """
    result = []

    def backtrack(current, remaining):
        if not remaining:
            result.append(current[:])
            return

        for i in range(len(remaining)):
            current.append(remaining[i])             # Choose
            backtrack(current, remaining[:i] + remaining[i+1:])  # Explore
            current.pop()                             # Unchoose

    backtrack([], nums)
    return result

nums = [1, 2, 3]
result = permutations(nums)
print(f"  Permutations of {nums}:")
for p in result:
    print(f"    {p}")
print(f"  Total: {len(result)} permutations ({len(nums)}! = {len(result)})")
````

````python
# Backtracking: Combination Sum
print("\nBACKTRACKING: Combination Sum")
print("=" * 60)

def combination_sum(candidates, target):
    """
    Find all combinations that sum to target (can reuse elements).
    Time: O(2^t) where t = target/min_candidate
    """
    result = []

    def backtrack(start, current, remaining):
        if remaining == 0:
            result.append(current[:])
            return
        if remaining < 0:
            return

        for i in range(start, len(candidates)):
            current.append(candidates[i])
            backtrack(i, current, remaining - candidates[i])  # i, not i+1 (reuse OK)
            current.pop()

    backtrack(0, [], target)
    return result

candidates = [2, 3, 6, 7]
target = 7
result = combination_sum(candidates, target)
print(f"  Candidates: {candidates}, Target: {target}")
print(f"  Combinations: {result}")
````

# Section 3: Introduction to Dynamic Programming

## What is Dynamic Programming (DP)?

DP = **Recursion + Caching** (avoid solving the same subproblem twice)

## When to Use DP

A problem has **optimal substructure** + **overlapping subproblems**:
1. **Optimal substructure**: The optimal solution can be built from optimal solutions to subproblems
2. **Overlapping subproblems**: Same subproblems are solved repeatedly

## How to Spot a DP Problem

Look for these keywords:
- "Find the **minimum/maximum** ..."
- "Count the **number of ways** ..."
- "Is it **possible** to ..."
- "Find the **longest/shortest** ..."

## Two Approaches

| Approach | Method | Direction |
|----------|--------|-----------|
| **Top-down (Memoization)** | Recursion + cache | Start from big problem, break down |
| **Bottom-up (Tabulation)** | Iterative + table | Start from smallest, build up |

Both give the same answer. Top-down is easier to write, bottom-up is often faster.

## The Classic Example: Fibonacci

```
fib(5) = fib(4) + fib(3)
fib(4) = fib(3) + fib(2)   <- fib(3) computed TWICE!
fib(3) = fib(2) + fib(1)   <- fib(2) computed THREE times!

Without caching: O(2^n) -- exponential
With caching:    O(n) -- each subproblem solved once
```

````python
# Fibonacci: Naive vs Memoized vs Tabulation
import time

print("DYNAMIC PROGRAMMING: Fibonacci")
print("=" * 60)

# Approach 1: Naive recursion -- O(2^n) TERRIBLE
def fib_naive(n):
    if n <= 1:
        return n
    return fib_naive(n - 1) + fib_naive(n - 2)

# Approach 2: Top-down (memoization) -- O(n)
def fib_memo(n, memo={}):
    if n <= 1:
        return n
    if n in memo:
        return memo[n]
    memo[n] = fib_memo(n - 1, memo) + fib_memo(n - 2, memo)
    return memo[n]

# Approach 3: Bottom-up (tabulation) -- O(n) time, O(n) space
def fib_tab(n):
    if n <= 1:
        return n
    dp = [0] * (n + 1)
    dp[1] = 1
    for i in range(2, n + 1):
        dp[i] = dp[i-1] + dp[i-2]
    return dp[n]

# Approach 4: Space-optimized -- O(n) time, O(1) space
def fib_optimal(n):
    if n <= 1:
        return n
    a, b = 0, 1
    for _ in range(2, n + 1):
        a, b = b, a + b
    return b

# Compare performance
print("  Comparing approaches:")
for n in [10, 20, 30]:
    start = time.perf_counter()
    r1 = fib_naive(n)
    t1 = time.perf_counter() - start

    start = time.perf_counter()
    r2 = fib_memo(n, {})
    t2 = time.perf_counter() - start

    print(f"  fib({n}) = {r1}")
    print(f"    Naive: {t1:.6f}s  |  Memoized: {t2:.8f}s")
    print()

# Large n (only possible with DP)
print(f"  fib(100) = {fib_optimal(100)}")
print("  (Naive recursion would take billions of years for fib(100)!)")
````

# Section 4: Classic DP Problems

## The DP Problem-Solving Framework

1. **Define the state**: What does `dp[i]` (or `dp[i][j]`) represent?
2. **Find the recurrence**: How does `dp[i]` relate to smaller subproblems?
3. **Identify base cases**: What are the trivial cases?
4. **Determine order**: In what order should we fill the table?
5. **Optimize space** (optional): Can we reduce from O(n) table to O(1)?

````python
# DP Problem 1: Climbing Stairs
print("DP PROBLEM 1: Climbing Stairs")
print("=" * 60)

def climb_stairs(n):
    """
    You can climb 1 or 2 steps. How many ways to reach the top?
    State: dp[i] = number of ways to reach step i
    Recurrence: dp[i] = dp[i-1] + dp[i-2]  (arrive from 1 step back OR 2 steps back)
    Base: dp[0] = 1, dp[1] = 1
    Time: O(n), Space: O(1)
    """
    if n <= 2:
        return n
    a, b = 1, 2
    for _ in range(3, n + 1):
        a, b = b, a + b
    return b

print("  How many ways to climb n stairs (1 or 2 steps at a time)?")
for n in range(1, 8):
    print(f"    climb_stairs({n}) = {climb_stairs(n)}")
print("\n  This is just Fibonacci in disguise!")
print("  dp[n] = dp[n-1] + dp[n-2]")
````

````python
# DP Problem 2: Coin Change
print("\nDP PROBLEM 2: Coin Change (Minimum Coins)")
print("=" * 60)

def coin_change(coins, amount):
    """
    Find minimum number of coins to make the amount.
    State: dp[i] = min coins needed for amount i
    Recurrence: dp[i] = min(dp[i - coin] + 1) for each coin
    Base: dp[0] = 0
    Time: O(amount * len(coins)), Space: O(amount)
    """
    dp = [float("inf")] * (amount + 1)
    dp[0] = 0

    for i in range(1, amount + 1):
        for coin in coins:
            if coin <= i and dp[i - coin] != float("inf"):
                dp[i] = min(dp[i], dp[i - coin] + 1)

    return dp[amount] if dp[amount] != float("inf") else -1

tests = [
    ([1, 5, 10, 25], 30, 2),    # 25 + 5
    ([1, 5, 10, 25], 11, 2),    # 10 + 1
    ([2], 3, -1),               # Impossible
    ([1, 2, 5], 11, 3),         # 5 + 5 + 1
]
for coins, amount, expected in tests:
    result = coin_change(coins, amount)
    status = "PASS" if result == expected else "FAIL"
    print(f"  [{status}] coins={coins}, amount={amount} -> {result} coins")

print("\n  Key insight: To make amount i, try each coin c:")
print("  If we use coin c, we need dp[i-c] + 1 coins. Take the minimum.")
````

````python
# DP Problem 3: Longest Common Subsequence (LCS)
print("\nDP PROBLEM 3: Longest Common Subsequence")
print("=" * 60)

def lcs(text1, text2):
    """
    Find length of longest common subsequence.
    State: dp[i][j] = LCS of text1[:i] and text2[:j]
    Recurrence:
      If text1[i-1] == text2[j-1]: dp[i][j] = dp[i-1][j-1] + 1
      Else: dp[i][j] = max(dp[i-1][j], dp[i][j-1])
    Time: O(m*n), Space: O(m*n)
    """
    m, n = len(text1), len(text2)
    dp = [[0] * (n + 1) for _ in range(m + 1)]

    for i in range(1, m + 1):
        for j in range(1, n + 1):
            if text1[i-1] == text2[j-1]:
                dp[i][j] = dp[i-1][j-1] + 1
            else:
                dp[i][j] = max(dp[i-1][j], dp[i][j-1])

    return dp[m][n]

tests = [
    ("abcde", "ace", 3),    # "ace"
    ("abc", "abc", 3),      # "abc"
    ("abc", "def", 0),      # no common
]
for t1, t2, expected in tests:
    result = lcs(t1, t2)
    status = "PASS" if result == expected else "FAIL"
    print(f"  [{status}] lcs("{t1}", "{t2}") = {result}")

print("\n  DE Connection: LCS is used in diff algorithms (git diff!)")
print("  Also: schema evolution detection -- what changed between versions?")
````

````python
# DP Problem 4: 0/1 Knapsack
print("\nDP PROBLEM 4: 0/1 Knapsack")
print("=" * 60)

def knapsack(weights, values, capacity):
    """
    Maximize value within weight capacity (each item used at most once).
    State: dp[i][w] = max value using first i items with capacity w
    Recurrence:
      Skip item i: dp[i][w] = dp[i-1][w]
      Take item i: dp[i][w] = dp[i-1][w - weights[i]] + values[i]
      dp[i][w] = max(skip, take)
    Time: O(n * capacity), Space: O(n * capacity)
    """
    n = len(weights)
    dp = [[0] * (capacity + 1) for _ in range(n + 1)]

    for i in range(1, n + 1):
        for w in range(capacity + 1):
            # Don't take item i
            dp[i][w] = dp[i-1][w]
            # Take item i (if it fits)
            if weights[i-1] <= w:
                dp[i][w] = max(dp[i][w], dp[i-1][w - weights[i-1]] + values[i-1])

    return dp[n][capacity]

weights = [2, 3, 4, 5]
values = [3, 4, 5, 6]
capacity = 8
result = knapsack(weights, values, capacity)
print(f"  Weights: {weights}")
print(f"  Values:  {values}")
print(f"  Capacity: {capacity}")
print(f"  Max value: {result}")
print("\n  At each item: do we take it or skip it? Classic binary decision.")
print("  DE Connection: Resource allocation with capacity constraints.")
````

````python
# DP Problem 5: Longest Increasing Subsequence (LIS)
print("\nDP PROBLEM 5: Longest Increasing Subsequence")
print("=" * 60)

def lis(nums):
    """
    Find length of longest strictly increasing subsequence.
    State: dp[i] = length of LIS ending at index i
    Recurrence: dp[i] = max(dp[j] + 1) for all j < i where nums[j] < nums[i]
    Time: O(n^2), Space: O(n)
    """
    if not nums:
        return 0
    n = len(nums)
    dp = [1] * n  # Each element is a subsequence of length 1

    for i in range(1, n):
        for j in range(i):
            if nums[j] < nums[i]:
                dp[i] = max(dp[i], dp[j] + 1)

    return max(dp)

tests = [
    ([10, 9, 2, 5, 3, 7, 101, 18], 4),  # [2, 3, 7, 101] or [2, 5, 7, 101]
    ([0, 1, 0, 3, 2, 3], 4),            # [0, 1, 2, 3]
    ([7, 7, 7, 7], 1),                   # [7]
]
for nums, expected in tests:
    result = lis(nums)
    status = "PASS" if result == expected else "FAIL"
    print(f"  [{status}] LIS({nums}) = {result}")
````

# Section 5: DP on Grids

Grid DP is very common in interviews. The idea:
- Start from one corner, reach another corner
- At each cell, you can go right or down (typically)
- Count paths, find minimum cost, etc.

````python
# Grid DP: Unique Paths
print("GRID DP: Unique Paths")
print("=" * 60)

def unique_paths(m, n):
    """
    Count paths from top-left to bottom-right (can only go right or down).
    State: dp[i][j] = number of paths to reach cell (i, j)
    Recurrence: dp[i][j] = dp[i-1][j] + dp[i][j-1]
    Time: O(m*n), Space: O(n)
    """
    dp = [1] * n  # First row: only one way (go right)

    for i in range(1, m):
        for j in range(1, n):
            dp[j] += dp[j-1]  # dp[j] = from above + from left

    return dp[-1]

print(f"  unique_paths(3, 7) = {unique_paths(3, 7)}")
print(f"  unique_paths(3, 3) = {unique_paths(3, 3)}")

# Minimum path sum
def min_path_sum(grid):
    """
    Find path from top-left to bottom-right with minimum sum.
    Time: O(m*n), Space: O(1) -- modify grid in place.
    """
    m, n = len(grid), len(grid[0])

    # Fill first row (can only come from left)
    for j in range(1, n):
        grid[0][j] += grid[0][j-1]
    # Fill first column (can only come from above)
    for i in range(1, m):
        grid[i][0] += grid[i-1][0]
    # Fill rest
    for i in range(1, m):
        for j in range(1, n):
            grid[i][j] += min(grid[i-1][j], grid[i][j-1])

    return grid[m-1][n-1]

grid = [[1, 3, 1], [1, 5, 1], [4, 2, 1]]
print(f"\n  min_path_sum([[1,3,1],[1,5,1],[4,2,1]]) = {min_path_sum([row[:] for row in grid])}")
print("  Path: 1 -> 3 -> 1 -> 1 -> 1 = 7")
````

````python
# SOLUTION 1: House Robber
print("SOLUTION 1: House Robber")
print("=" * 60)

def rob(nums):
    """
    dp[i] = max(dp[i-1], dp[i-2] + nums[i])
    Space-optimized to O(1).
    Time: O(n), Space: O(1)
    """
    if not nums:
        return 0
    if len(nums) == 1:
        return nums[0]

    prev2 = 0  # dp[i-2]
    prev1 = 0  # dp[i-1]

    for num in nums:
        current = max(prev1, prev2 + num)
        prev2 = prev1
        prev1 = current

    return prev1

tests = [
    ([1, 2, 3, 1], 4),
    ([2, 7, 9, 3, 1], 12),
    ([2, 1, 1, 2], 4),
]
for nums, expected in tests:
    result = rob(nums)
    status = "PASS" if result == expected else "FAIL"
    print(f"  [{status}] rob({nums}) = {result}")
````

````python
# SOLUTION 2: Word Break
print("\nSOLUTION 2: Word Break")
print("=" * 60)

def word_break(s, word_dict):
    """
    dp[i] = True if s[:i] can be segmented into words.
    Time: O(n^2 * m) where m = max word length, Space: O(n)
    """
    word_set = set(word_dict)
    n = len(s)
    dp = [False] * (n + 1)
    dp[0] = True  # Empty string can be segmented

    for i in range(1, n + 1):
        for j in range(i):
            if dp[j] and s[j:i] in word_set:
                dp[i] = True
                break

    return dp[n]

tests = [
    ("leetcode", ["leet", "code"], True),
    ("applepenapple", ["apple", "pen"], True),
    ("catsandog", ["cats", "dog", "sand", "and", "cat"], False),
]
for s, words, expected in tests:
    result = word_break(s, words)
    status = "PASS" if result == expected else "FAIL"
    print(f"  [{status}] word_break("{s}") = {result}")
````

````python
# SOLUTION 3: Edit Distance
print("\nSOLUTION 3: Edit Distance")
print("=" * 60)

def edit_distance(word1, word2):
    """
    Classic 2D DP.
    Time: O(m*n), Space: O(m*n)
    """
    m, n = len(word1), len(word2)
    dp = [[0] * (n + 1) for _ in range(m + 1)]

    # Base cases: transforming to/from empty string
    for i in range(m + 1):
        dp[i][0] = i  # Delete all characters
    for j in range(n + 1):
        dp[0][j] = j  # Insert all characters

    for i in range(1, m + 1):
        for j in range(1, n + 1):
            if word1[i-1] == word2[j-1]:
                dp[i][j] = dp[i-1][j-1]  # Characters match, no operation
            else:
                dp[i][j] = 1 + min(
                    dp[i-1][j],      # Delete from word1
                    dp[i][j-1],      # Insert into word1
                    dp[i-1][j-1]     # Replace
                )

    return dp[m][n]

tests = [
    ("horse", "ros", 3),
    ("intention", "execution", 5),
    ("", "abc", 3),
]
for w1, w2, expected in tests:
    result = edit_distance(w1, w2)
    status = "PASS" if result == expected else "FAIL"
    print(f"  [{status}] edit_distance("{w1}", "{w2}") = {result}")

print("\n  DE Connection: Edit distance is used in fuzzy matching!")
print("  Finding similar customer names, detecting typos in data.")
````

# Section 7: Summary

## How to Identify DP Problems

| Keyword | DP Type | Example |
|---------|---------|---------|
| "Minimum cost/steps" | Min/Max DP | Coin Change, Edit Distance |
| "Number of ways" | Counting DP | Climbing Stairs, Unique Paths |
| "Is it possible?" | Boolean DP | Word Break, Subset Sum |
| "Longest/shortest sequence" | Sequence DP | LCS, LIS |
| "Maximize profit/value" | Optimization DP | Knapsack, House Robber |

## DP Patterns Summary

| Pattern | State | Example |
|---------|-------|---------|
| **Linear** | dp[i] depends on dp[i-1], dp[i-2] | Fibonacci, Climbing Stairs, House Robber |
| **Grid** | dp[i][j] depends on adjacent cells | Unique Paths, Min Path Sum |
| **String** | dp[i][j] on two strings | LCS, Edit Distance |
| **Knapsack** | dp[i][w] on items + capacity | 0/1 Knapsack, Subset Sum |
| **Interval** | dp[i][j] on subarray i..j | Matrix Chain, Burst Balloons |
