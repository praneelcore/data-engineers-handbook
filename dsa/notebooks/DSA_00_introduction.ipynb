# Section 1: What is DSA?

## The Simple Explanation

**Data Structures** = Ways to organize and store data
**Algorithms** = Step-by-step instructions to solve a problem

Think of it like a kitchen:
- **Data structures** are your storage containers (bowls, shelves, fridge, spice rack)
- **Algorithms** are your recipes (step-by-step instructions to cook a dish)

A good chef knows which container to use for which ingredient, and which recipe
is fastest for the occasion. A good programmer knows which data structure to use
for which problem, and which algorithm is most efficient.

## Real-World Analogy

Imagine you have 1,000 books and need to find one specific book.

**Approach 1 (no structure):** Books scattered randomly on the floor.
You check each book one by one. Average: 500 checks. Worst case: 1,000 checks.

**Approach 2 (sorted + binary search):** Books sorted alphabetically on shelves.
You open to the middle, go left or right, repeat. Maximum: ~10 checks!

That's the power of DSA: the right structure + the right algorithm = dramatically faster.

## What DSA Is NOT

- It's NOT about memorizing code
- It's NOT just for competitive programming
- It's NOT only for "genius" programmers
- It's NOT irrelevant to data engineering

## What DSA IS

- A toolkit for solving problems efficiently
- A way to think about trade-offs (speed vs memory)
- The foundation of every software system you use
- What interviewers test at Google, Meta, Amazon, Databricks, Uber

````python
# The power of the right approach -- a concrete example
import time

# Problem: Find if a number exists in a collection of 1 million numbers

# Approach 1: List (unsorted) -- check every element
def search_list(data, target):
    """Search by checking every element. Slow for large data."""
    for item in data:
        if item == target:
            return True
    return False

# Approach 2: Set (hash table) -- instant lookup
def search_set(data, target):
    """Search using a set. Fast regardless of data size."""
    return target in data

# Create 1 million numbers
numbers_list = list(range(1_000_000))
numbers_set = set(numbers_list)
target = 999_999  # Near the end -- worst case for list

# Time both approaches
print("Finding 999,999 in 1,000,000 numbers:")
print("=" * 50)

start = time.perf_counter()
result = search_list(numbers_list, target)
list_time = time.perf_counter() - start
print(f"List search:  {list_time:.6f} seconds  (found: {result})")

start = time.perf_counter()
result = search_set(numbers_set, target)
set_time = time.perf_counter() - start
print(f"Set lookup:   {set_time:.6f} seconds  (found: {result})")

if set_time > 0:
    speedup = list_time / set_time
    print(f"\nSpeedup: {speedup:,.0f}x faster with the right data structure!")

print("\nSame problem, same result, wildly different speed.")
print("THIS is why DSA matters.")
````

````python
# DSA in data engineering -- concrete examples
print("DSA IN DATA ENGINEERING -- Real Examples")
print("=" * 60)

# Example 1: GROUP BY uses a hash map
print("\n1. GROUP BY = Hash Map")
orders = [
    {"customer": "Alice", "amount": 100},
    {"customer": "Bob", "amount": 200},
    {"customer": "Alice", "amount": 150},
    {"customer": "Charlie", "amount": 300},
    {"customer": "Bob", "amount": 75},
]

# This is exactly what SQL GROUP BY does internally:
revenue_by_customer = {}
for order in orders:
    customer = order["customer"]
    revenue_by_customer[customer] = revenue_by_customer.get(customer, 0) + order["amount"]

print(f"   Revenue by customer: {revenue_by_customer}")
print(f"   Time complexity: O(n) -- one pass through the data")

# Example 2: Deduplication uses a hash set
print("\n2. DEDUPLICATION = Hash Set")
raw_ids = [1, 2, 3, 2, 4, 1, 5, 3, 6]
unique_ids = list(set(raw_ids))
print(f"   Raw IDs: {raw_ids}")
print(f"   Unique IDs (via set): {sorted(unique_ids)}")
print(f"   Time complexity: O(n)")

# Example 3: Top-N uses a heap
import heapq
print("\n3. TOP-N CUSTOMERS = Heap (Priority Queue)")
customers = [
    ("Alice", 5000), ("Bob", 12000), ("Charlie", 3000),
    ("Diana", 8000), ("Eve", 15000), ("Frank", 6000),
]
top_3 = heapq.nlargest(3, customers, key=lambda x: x[1])
print(f"   All customers: {[(c, f'${v:,}') for c, v in customers]}")
print(f"   Top 3 by revenue: {[(c, f'${v:,}') for c, v in top_3]}")
print(f"   Time complexity: O(n log k) -- better than sorting all")

# Example 4: Pipeline ordering uses topological sort
print("\n4. PIPELINE ORDERING = Topological Sort (Graph)")
dependencies = {
    "ingest_raw": [],
    "clean_data": ["ingest_raw"],
    "transform": ["clean_data"],
    "build_gold": ["transform"],
    "quality_check": ["build_gold"],
}

def simple_topo_sort(deps):
    visited = set()
    order = []
    def visit(node):
        if node in visited:
            return
        visited.add(node)
        for dep in deps.get(node, []):
            visit(dep)
        order.append(node)
    for node in deps:
        visit(node)
    return order

execution_order = simple_topo_sort(dependencies)
print(f"   Execution order: {execution_order}")
print(f"   This is exactly what Airflow/Databricks Workflows does!")
````

````python
# Python refresher: Classes (needed for linked lists, trees, graphs)
print("CLASSES -- Custom data structures")
print("=" * 50)

class Node:
    """A simple node with a value and a next pointer."""
    def __init__(self, value):
        self.value = value
        self.next = None

    def __repr__(self):
        return f"Node({self.value})"

# Create nodes
node1 = Node(10)
node2 = Node(20)
node3 = Node(30)

# Link them together (this is a linked list!)
node1.next = node2
node2.next = node3

# Traverse
print("Linked list traversal:")
current = node1
while current:
    print(f"  {current.value}", end="")
    if current.next:
        print(" -> ", end="")
    current = current.next
print(" -> None")

print("\nYou just built your first linked list!")
print("We will cover this in depth in DSA_03.")
````

# Section 4: The Core Data Structures -- Overview

There are about 10 fundamental data structures. Everything else is built from these.

```
DATA STRUCTURES
|
+-- Linear (elements in a sequence)
|   +-- Array / List      -- indexed, fast access by position
|   +-- Linked List       -- nodes connected by pointers
|   +-- Stack             -- Last In, First Out (LIFO)
|   +-- Queue             -- First In, First Out (FIFO)
|
+-- Non-Linear (elements with relationships)
|   +-- Tree              -- hierarchical (parent-child)
|   |   +-- Binary Tree
|   |   +-- Binary Search Tree (BST)
|   |   +-- Heap (priority queue)
|   +-- Graph             -- nodes connected by edges
|
+-- Hash-Based
    +-- Hash Map / Dict   -- key-value pairs, O(1) lookup
    +-- Hash Set          -- unique values, O(1) lookup
```

## When to Use Each

| Data Structure | Use When | Python Equivalent |
|---------------|----------|-------------------|
| **Array/List** | Ordered data, index access | `list` |
| **Hash Map** | Fast key-value lookup | `dict` |
| **Hash Set** | Fast membership check, unique values | `set` |
| **Stack** | Undo/redo, bracket matching, DFS | `list` (append/pop) |
| **Queue** | Task scheduling, BFS | `collections.deque` |
| **Heap** | Always need min/max quickly | `heapq` module |
| **Tree** | Hierarchical data, sorted data | Custom class |
| **Graph** | Networks, relationships, dependencies | `dict` of lists |

````python
# Quick demo of each core data structure
from collections import deque
import heapq

print("CORE DATA STRUCTURES IN PYTHON")
print("=" * 60)

# 1. List (Array)
print("\n1. LIST (Array) -- ordered, indexed")
my_list = [10, 20, 30, 40, 50]
print(f"   Data: {my_list}")
print(f"   Access by index: my_list[2] = {my_list[2]}")
print(f"   Append: O(1), Search: O(n)")

# 2. Dictionary (Hash Map)
print("\n2. DICT (Hash Map) -- key-value pairs")
my_dict = {"alice": 95, "bob": 87, "charlie": 92}
print(f"   Data: {my_dict}")
print(f"   Lookup: my_dict['alice'] = {my_dict['alice']}")
print(f"   Lookup: O(1) average -- instant!")

# 3. Set (Hash Set)
print("\n3. SET (Hash Set) -- unique values")
my_set = {1, 2, 3, 4, 5}
print(f"   Data: {my_set}")
print(f"   Membership: 3 in my_set = {3 in my_set}")
print(f"   Membership: O(1) average -- instant!")

# 4. Stack (using list)
print("\n4. STACK -- Last In, First Out (LIFO)")
stack = []
stack.append("first")
stack.append("second")
stack.append("third")
print(f"   After pushing 3 items: {stack}")
top = stack.pop()
print(f"   After pop: removed '{top}', stack = {stack}")

# 5. Queue (using deque)
print("\n5. QUEUE -- First In, First Out (FIFO)")
queue = deque()
queue.append("first")
queue.append("second")
queue.append("third")
print(f"   After enqueuing 3 items: {list(queue)}")
front = queue.popleft()
print(f"   After dequeue: removed '{front}', queue = {list(queue)}")

# 6. Heap (Priority Queue)
print("\n6. HEAP -- always get min quickly")
heap = [5, 3, 8, 1, 9, 2]
heapq.heapify(heap)
print(f"   After heapify: {heap}")
print(f"   Get minimum: {heap[0]}")
smallest = heapq.heappop(heap)
print(f"   After pop: removed {smallest}, heap = {heap}")
````

# Section 5: Your First DSA Problems

## How to Approach a DSA Problem

Use the **UMPIRE** framework:

```
U -- Understand the problem (read carefully, clarify inputs/outputs)
M -- Match to a pattern (does this look like something you've seen?)
P -- Plan your approach (think before coding)
I -- Implement (write clean code)
R -- Review (test with examples and edge cases)
E -- Evaluate (time/space complexity, can you do better?)
```

Let's solve 3 problems together using this framework.

## Problem 1: Find the Maximum Element

**Problem:** Given a list of numbers, find the largest one.

This is trivial, but let's practice the UMPIRE framework on it.

````python
# Problem 1: Find the Maximum Element
print("PROBLEM 1: Find the Maximum Element")
print("=" * 50)

# U -- Understand
print("Understanding:")
print("  Input: a list of numbers (could be positive, negative, or mixed)")
print("  Output: the single largest number")
print("  Edge cases: empty list? single element?")

# M -- Match
print("\nMatch: This is a simple linear scan problem")

# P -- Plan
print("\nPlan:")
print("  1. Start with the first element as our 'max so far'")
print("  2. Go through each element")
print("  3. If it's bigger than our current max, update")
print("  4. Return the max")

# I -- Implement
def find_max(arr):
    """Find the maximum element in a list."""
    if not arr:
        return None  # Edge case: empty list

    max_val = arr[0]
    for num in arr[1:]:
        if num > max_val:
            max_val = num
    return max_val

# R -- Review
print("\nReview (testing):")
test_cases = [
    ([3, 1, 4, 1, 5, 9, 2, 6], 9),
    ([-5, -2, -8, -1], -1),
    ([42], 42),
    ([7, 7, 7, 7], 7),
]
for arr, expected in test_cases:
    result = find_max(arr)
    status = "PASS" if result == expected else "FAIL"
    print(f"  [{status}] find_max({arr}) = {result}")

# E -- Evaluate
print("\nEvaluate:")
print("  Time: O(n) -- we look at each element once")
print("  Space: O(1) -- only one variable (max_val)")
print("  Can we do better? No! Must look at every element at least once.")
````

## Problem 2: Two Sum

**Problem:** Given an array of integers and a target sum,
return the indices of two numbers that add up to the target.

```
Input:  nums = [2, 7, 11, 15], target = 9
Output: [0, 1]  (because nums[0] + nums[1] = 2 + 7 = 9)
```

This is THE most famous interview problem. Let's solve it two ways.

````python
# Problem 2: Two Sum -- Brute Force first
print("PROBLEM 2: Two Sum")
print("=" * 50)

# APPROACH 1: Brute Force -- check every pair
def two_sum_brute(nums, target):
    """Check every pair. O(n^2) time, O(1) space."""
    for i in range(len(nums)):
        for j in range(i + 1, len(nums)):
            if nums[i] + nums[j] == target:
                return [i, j]
    return []

print("Approach 1: Brute Force O(n^2)")
nums = [2, 7, 11, 15]
result = two_sum_brute(nums, 9)
print(f"  two_sum({nums}, 9) = {result}")
print(f"  Verification: nums[{result[0]}] + nums[{result[1]}] = {nums[result[0]]} + {nums[result[1]]} = {nums[result[0]] + nums[result[1]]}")
print(f"  Time: O(n^2) -- two nested loops")
print(f"  Space: O(1) -- no extra storage")
````

````python
# Problem 2: Two Sum -- Optimized with Hash Map
print("\nApproach 2: Hash Map O(n)")
print("-" * 50)

def two_sum(nums, target):
    """Use hash map for O(1) complement lookup. O(n) time, O(n) space."""
    seen = {}  # {number: index}
    for i, num in enumerate(nums):
        complement = target - num
        if complement in seen:
            return [seen[complement], i]
        seen[num] = i
    return []

# Walk through the logic
print("How it works:")
print("  For each number, we ask: 'Does (target - number) exist in what I've seen?'")
print()
nums = [2, 7, 11, 15]
target = 9
print(f"  nums = {nums}, target = {target}")
print(f"  Step 1: num=2, need 9-2=7, seen={{}} -> not found, add 2")
print(f"  Step 2: num=7, need 9-7=2, seen={{2:0}} -> FOUND at index 0!")
print(f"  Return [0, 1]")
print()

# Test
test_cases = [
    ([2, 7, 11, 15], 9, [0, 1]),
    ([3, 2, 4], 6, [1, 2]),
    ([3, 3], 6, [0, 1]),
]
for nums, target, expected in test_cases:
    result = two_sum(nums, target)
    status = "PASS" if result == expected else "FAIL"
    print(f"  [{status}] two_sum({nums}, {target}) = {result}")

print("\n  Time: O(n) -- single pass through array")
print("  Space: O(n) -- hash map stores up to n elements")
print("  Trade-off: We used more memory to get faster time!")
````

## Problem 3: Contains Duplicate

**Problem:** Given an array, return True if any value appears at least twice.

```
Input:  [1, 2, 3, 1]  -> True  (1 appears twice)
Input:  [1, 2, 3, 4]  -> False (all unique)
```

````python
# Problem 3: Contains Duplicate
print("PROBLEM 3: Contains Duplicate")
print("=" * 50)

# Approach 1: Brute force -- check every pair
def contains_dup_brute(nums):
    """O(n^2) -- compare every pair."""
    for i in range(len(nums)):
        for j in range(i + 1, len(nums)):
            if nums[i] == nums[j]:
                return True
    return False

# Approach 2: Sort first, check neighbors
def contains_dup_sort(nums):
    """O(n log n) -- sort, then adjacent duplicates are next to each other."""
    sorted_nums = sorted(nums)
    for i in range(len(sorted_nums) - 1):
        if sorted_nums[i] == sorted_nums[i + 1]:
            return True
    return False

# Approach 3: Hash set
def contains_dup_set(nums):
    """O(n) -- if we see a number we've already added to the set, it's a duplicate."""
    seen = set()
    for num in nums:
        if num in seen:
            return True
        seen.add(num)
    return False

# Compare all three
test_cases = [
    ([1, 2, 3, 1], True),
    ([1, 2, 3, 4], False),
    ([1, 1, 1, 3, 3, 4, 3, 2, 4, 2], True),
]

print("Testing all three approaches:")
for nums, expected in test_cases:
    r1 = contains_dup_brute(nums)
    r2 = contains_dup_sort(nums)
    r3 = contains_dup_set(nums)
    assert r1 == r2 == r3 == expected
    print(f"  {nums} -> {r3} (all approaches agree)")

print("\nComparison:")
print(f"  {'Approach':<20} {'Time':<15} {'Space':<10}")
print(f"  {'-'*45}")
print(f"  {'Brute force':<20} {'O(n^2)':<15} {'O(1)':<10}")
print(f"  {'Sort first':<20} {'O(n log n)':<15} {'O(n)':<10}")
print(f"  {'Hash set':<20} {'O(n)':<15} {'O(n)':<10}")
print("\n  Winner: Hash set -- fastest time, reasonable space.")
````

# Solutions

````python
# SOLUTION A: Reverse a List
def reverse_list(arr):
    """Reverse in-place using two pointers. O(n) time, O(1) space."""
    arr = arr[:]  # Make a copy to not modify original
    left, right = 0, len(arr) - 1
    while left < right:
        arr[left], arr[right] = arr[right], arr[left]
        left += 1
        right -= 1
    return arr

# Test
print("Problem A: Reverse a List")
tests = [
    ([1, 2, 3, 4, 5], [5, 4, 3, 2, 1]),
    ([1], [1]),
    ([], []),
    ([1, 2], [2, 1]),
]
for arr, expected in tests:
    result = reverse_list(arr)
    status = "PASS" if result == expected else "FAIL"
    print(f"  [{status}] reverse_list({arr}) = {result}")
````

````python
# SOLUTION B: Count Character Frequency
def char_frequency(s):
    """Count each character. O(n) time, O(k) space where k = unique chars."""
    freq = {}
    for char in s:
        freq[char] = freq.get(char, 0) + 1
    return freq

# Test
print("Problem B: Character Frequency")
tests = [
    ("hello", {"h": 1, "e": 1, "l": 2, "o": 1}),
    ("aaa", {"a": 3}),
    ("", {}),
]
for s, expected in tests:
    result = char_frequency(s)
    status = "PASS" if result == expected else "FAIL"
    print(f"  [{status}] char_frequency(\"{s}\") = {result}")
````

````python
# SOLUTION C: Find Missing Number
def find_missing(nums):
    """Use math: expected_sum - actual_sum = missing number. O(n) time, O(1) space."""
    n = len(nums)
    expected_sum = n * (n + 1) // 2
    actual_sum = sum(nums)
    return expected_sum - actual_sum

# Test
print("Problem C: Find Missing Number")
tests = [
    ([3, 0, 1], 2),
    ([0, 1], 2),
    ([9, 6, 4, 2, 3, 5, 7, 0, 1], 8),
    ([0], 1),
]
for nums, expected in tests:
    result = find_missing(nums)
    status = "PASS" if result == expected else "FAIL"
    print(f"  [{status}] find_missing({nums}) = {result}")
````
