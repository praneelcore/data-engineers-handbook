# Section 1: What is Big-O Notation?

## The Core Idea

Big-O answers one question: **"As my input grows, how does the work grow?"**

We don't care about:
- The exact number of operations
- The speed of your computer
- Constants and small terms

We care about: **the growth RATE as n gets very large.**

## Analogy: Mailing Packages

Imagine you need to send data to a friend:

| Method | 1 MB | 1 GB | 1 TB |
|--------|------|------|------|
| **Internet (O(n))** | 1 sec | 17 min | 12 days |
| **FedEx a hard drive (O(1))** | 24 hrs | 24 hrs | 24 hrs |

For small data, Internet wins. For HUGE data, FedEx wins.
This is why Big-O matters: it tells you which approach scales better.

## The Common Complexity Classes

| Big-O | Name | Example | 1M items |
|-------|------|---------|----------|
| O(1) | Constant | dict lookup | instant |
| O(log n) | Logarithmic | Binary search | ~20 steps |
| O(n) | Linear | Single loop | 1M steps |
| O(n log n) | Linearithmic | Merge sort | ~20M steps |
| O(n^2) | Quadratic | Nested loops | 1 TRILLION steps |
| O(2^n) | Exponential | All subsets | heat death of universe |
| O(n!) | Factorial | All permutations | forget it |

## The Key Insight

> O(n^2) with n=1000 = 1,000,000 operations (fine)
> O(n^2) with n=1,000,000 = 1,000,000,000,000 operations (HOURS)
> O(n) with n=1,000,000 = 1,000,000 operations (milliseconds)

**The gap between O(n) and O(n^2) is the difference between
"runs instantly" and "never finishes" for large data.**

````python
# Visualizing growth rates
print("GROWTH RATES -- How operations scale with input size")
print("=" * 65)
print()
print(f"{'n':<10} {'O(1)':<10} {'O(log n)':<10} {'O(n)':<12} {'O(n log n)':<14} {'O(n^2)':<14} {'O(2^n)':<14}")
print("-" * 74)

import math

for n in [10, 100, 1000, 10000, 100000, 1000000]:
    o1 = 1
    ologn = int(math.log2(n)) if n > 0 else 0
    on = n
    onlogn = int(n * math.log2(n))
    on2 = n * n
    o2n = "HUGE" if n > 30 else str(2**n)

    print(f"{n:<10} {o1:<10} {ologn:<10} {on:<12,} {onlogn:<14,} {on2:<14,} {o2n:<14}")

print()
print("Notice: O(n^2) at n=1,000,000 is 1 TRILLION operations.")
print("At 1 billion ops/second, that takes ~17 MINUTES.")
print("O(n) at n=1,000,000 takes ~1 MILLISECOND.")
````

# Section 2: Each Complexity Class Explained

## O(1) -- Constant Time

The operation takes the same time regardless of input size.

**Examples:**
- Accessing an array element by index: `arr[5]`
- Looking up a key in a dictionary: `d["key"]`
- Checking if a set contains a value: `x in my_set`
- Push/pop from a stack
- Getting the length of a list: `len(arr)`

````python
# O(1) -- Constant time examples
import time

print("O(1) -- CONSTANT TIME")
print("=" * 50)
print("Operation takes same time regardless of input size:")
print()

# Array access is O(1)
for size in [100, 10_000, 1_000_000, 10_000_000]:
    arr = list(range(size))
    start = time.perf_counter()
    _ = arr[size // 2]  # Access middle element
    elapsed = time.perf_counter() - start
    print(f"  arr[{size//2:>8,}] in array of {size:>10,}: {elapsed:.8f}s")

print()
print("  All roughly the same time -- that is O(1)!")

# Dict lookup is O(1)
print()
print("  Dict lookup is also O(1):")
for size in [100, 10_000, 1_000_000]:
    d = {i: i*2 for i in range(size)}
    start = time.perf_counter()
    _ = d[size // 2]
    elapsed = time.perf_counter() - start
    print(f"  dict[{size//2:>8,}] in dict of {size:>10,}: {elapsed:.8f}s")
````

## O(log n) -- Logarithmic Time

Each step cuts the problem in HALF. Very efficient.

**Examples:**
- Binary search in a sorted array
- Finding an element in a balanced BST
- Calculating power using divide and conquer

**Intuition:** log2(1,000,000) = ~20. So even with 1 million items,
you only need ~20 steps!

````python
# O(log n) -- Logarithmic time
print("O(log n) -- LOGARITHMIC TIME")
print("=" * 50)
print("Each step cuts the problem in half:")
print()

def binary_search(arr, target):
    """Find target in sorted array. O(log n) time."""
    left, right = 0, len(arr) - 1
    steps = 0
    while left <= right:
        steps += 1
        mid = (left + right) // 2
        if arr[mid] == target:
            return mid, steps
        elif arr[mid] < target:
            left = mid + 1
        else:
            right = mid - 1
    return -1, steps

# Show how steps grow SLOWLY even as input grows fast
print(f"  {'Array Size':<15} {'Steps to Find':<15} {'log2(n)':<10}")
print(f"  {'-'*40}")

for size in [10, 100, 1000, 10000, 100000, 1000000, 10000000]:
    arr = list(range(size))
    target = size - 1  # Worst case: last element
    _, steps = binary_search(arr, target)
    log_n = int(math.log2(size))
    print(f"  {size:<15,} {steps:<15} {log_n:<10}")

print()
print("  10 MILLION items -- only ~24 steps!")
print("  Compare to O(n) which would need 10,000,000 steps.")
````

## O(n) -- Linear Time

You look at every element exactly once (or a constant number of times).

**Examples:**
- Finding max/min in an array
- Counting elements
- Single for loop over the data
- Linear search

````python
# O(n) -- Linear time
print("O(n) -- LINEAR TIME")
print("=" * 50)
print("Time grows proportionally with input size:")
print()

def find_max(arr):
    """Find maximum element. O(n) -- must check every element."""
    max_val = arr[0]
    for num in arr[1:]:
        if num > max_val:
            max_val = num
    return max_val

def sum_all(arr):
    """Sum all elements. O(n) -- visit each once."""
    total = 0
    for num in arr:
        total += num
    return total

# Time it for different sizes
print(f"  {'Size':<12} {'find_max time':<18} {'sum_all time':<18}")
print(f"  {'-'*48}")

for size in [10_000, 100_000, 1_000_000, 10_000_000]:
    arr = list(range(size))

    start = time.perf_counter()
    find_max(arr)
    t1 = time.perf_counter() - start

    start = time.perf_counter()
    sum_all(arr)
    t2 = time.perf_counter() - start

    print(f"  {size:<12,} {t1:<18.6f} {t2:<18.6f}")

print()
print("  Notice: 10x more data = ~10x more time. That is linear growth.")
````

## O(n log n) -- Linearithmic Time

Slightly worse than linear, but much better than quadratic.
This is the best possible time for comparison-based sorting.

**Examples:**
- Merge sort, Quick sort (average), Heap sort
- Python's built-in `sorted()` (Timsort)
- Many divide-and-conquer algorithms

````python
# O(n log n) -- Linearithmic time
import random

print("O(n log n) -- LINEARITHMIC TIME")
print("=" * 50)
print("This is how fast sorting can be:")
print()

print(f"  {'Size':<12} {'sorted() time':<18} {'Ratio':<10}")
print(f"  {'-'*40}")

prev_time = None
for size in [10_000, 100_000, 1_000_000, 5_000_000]:
    arr = [random.randint(0, size) for _ in range(size)]

    start = time.perf_counter()
    sorted(arr)
    elapsed = time.perf_counter() - start

    ratio = f"{elapsed/prev_time:.1f}x" if prev_time else "-"
    print(f"  {size:<12,} {elapsed:<18.6f} {ratio:<10}")
    prev_time = elapsed

print()
print("  If it were O(n), 10x data = 10x time.")
print("  If it were O(n^2), 10x data = 100x time.")
print("  O(n log n) is between: 10x data = ~13x time (for large n).")
````

## O(n^2) -- Quadratic Time

Usually means nested loops where both iterate over n.
Gets painfully slow for large inputs.

**Examples:**
- Bubble sort, Selection sort, Insertion sort
- Checking every pair of elements
- Naive duplicate detection

````python
# O(n^2) -- Quadratic time
print("O(n^2) -- QUADRATIC TIME")
print("=" * 50)
print("Nested loops = quadratic. Gets slow FAST:")
print()

def has_duplicate_naive(arr):
    """Check every pair for duplicates. O(n^2)."""
    n = len(arr)
    for i in range(n):
        for j in range(i + 1, n):
            if arr[i] == arr[j]:
                return True
    return False

print(f"  {'Size':<10} {'Time':<15} {'Ratio':<10}")
print(f"  {'-'*35}")

prev_time = None
for size in [1000, 2000, 4000, 8000, 16000]:
    arr = list(range(size))  # No duplicates = worst case

    start = time.perf_counter()
    has_duplicate_naive(arr)
    elapsed = time.perf_counter() - start

    ratio = f"{elapsed/prev_time:.1f}x" if prev_time else "-"
    print(f"  {size:<10,} {elapsed:<15.4f} {ratio:<10}")
    prev_time = elapsed

print()
print("  Double the input -> ~4x the time. That is quadratic growth.")
print("  At n=1,000,000 this would take HOURS.")
print()
print("  FIX: Use a set for O(n) duplicate detection:")
print("    def has_duplicate(arr): return len(arr) != len(set(arr))")
````

## O(2^n) and O(n!) -- Exponential and Factorial

These are only acceptable for tiny inputs (n < 20-25).

**Examples:**
- Generating all subsets of a set: O(2^n)
- Generating all permutations: O(n!)
- Naive recursive Fibonacci: O(2^n)
- Brute-force traveling salesman: O(n!)

````python
# O(2^n) -- Exponential time
print("O(2^n) -- EXPONENTIAL TIME")
print("=" * 50)
print("Doubles with each additional element. Only viable for n < ~25:")
print()

def all_subsets(arr):
    """Generate all subsets. O(2^n) -- each element is in/out."""
    if not arr:
        return [[]]
    rest = all_subsets(arr[1:])
    return rest + [[arr[0]] + s for s in rest]

print(f"  {'n':<5} {'Subsets':<12} {'Time':<12}")
print(f"  {'-'*30}")

for n in [5, 10, 15, 18, 20, 22]:
    arr = list(range(n))
    start = time.perf_counter()
    result = all_subsets(arr)
    elapsed = time.perf_counter() - start
    print(f"  {n:<5} {len(result):<12,} {elapsed:<12.4f}")
    if elapsed > 2:
        print("  ... stopping (getting too slow)")
        break

print()
print("  n=20: over 1 million subsets.")
print("  n=30 would be over 1 billion -- completely impractical.")
````

# Section 3: How to Analyze Code

## Rules for Determining Big-O

### Rule 1: Drop constants
- O(2n) = O(n)
- O(100n) = O(n)
- O(n/2) = O(n)

### Rule 2: Drop lower-order terms
- O(n^2 + n) = O(n^2)
- O(n + log n) = O(n)
- O(n^3 + n^2 + n) = O(n^3)

### Rule 3: Different inputs = different variables
- If you have array A (size m) and array B (size n): O(m + n), not O(2n)

### Rule 4: Multiplication for nested operations
- Loop inside a loop: O(n) * O(n) = O(n^2)
- Loop inside a loop inside a loop: O(n^3)

### Rule 5: Addition for sequential operations
- One loop AFTER another: O(n) + O(n) = O(2n) = O(n)

````python
# Analyzing code -- examples
print("HOW TO ANALYZE CODE -- Examples")
print("=" * 60)

# Example 1: O(1)
print("\nExample 1: O(1)")
print("  def get_first(arr):")
print("      return arr[0]")
print("  -> One operation regardless of array size = O(1)")

# Example 2: O(n)
print("\nExample 2: O(n)")
print("  def sum_all(arr):")
print("      total = 0")
print("      for x in arr:        # n iterations")
print("          total += x       # O(1) per iteration")
print("      return total")
print("  -> n iterations * O(1) each = O(n)")

# Example 3: O(n^2)
print("\nExample 3: O(n^2)")
print("  def all_pairs(arr):")
print("      for i in range(len(arr)):       # n iterations")
print("          for j in range(len(arr)):   # n iterations each")
print("              print(arr[i], arr[j])   # O(1)")
print("  -> n * n * O(1) = O(n^2)")

# Example 4: O(n + m) -- two different inputs
print("\nExample 4: O(n + m)")
print("  def merge_lists(a, b):")
print("      for x in a:    # O(n) where n = len(a)")
print("          print(x)")
print("      for y in b:    # O(m) where m = len(b)")
print("          print(y)")
print("  -> O(n) + O(m) = O(n + m)")

# Example 5: O(n log n)
print("\nExample 5: O(n log n)")
print("  def sort_and_find(arr):")
print("      arr.sort()                 # O(n log n)")
print("      return binary_search(arr)  # O(log n)")
print("  -> O(n log n) + O(log n) = O(n log n)  [drop lower term]")
````

````python
# Practice: What is the Big-O?
print("PRACTICE: What is the Big-O of each function?")
print("=" * 60)

# Function 1
def func1(arr):
    total = 0
    for x in arr:
        total += x
    for x in arr:
        total += x * 2
    return total

# Function 2
def func2(arr):
    for i in range(len(arr)):
        for j in range(i, len(arr)):
            pass

# Function 3
def func3(n):
    i = n
    count = 0
    while i > 1:
        i = i // 2
        count += 1
    return count

# Function 4
def func4(arr):
    if len(arr) <= 1:
        return arr
    mid = len(arr) // 2
    left = func4(arr[:mid])
    right = func4(arr[mid:])
    return left + right  # simplified merge

print()
print("Function 1: Two sequential loops over n")
print("  Answer: O(n) + O(n) = O(2n) = O(n)")
print()
print("Function 2: Nested loop, j starts at i")
print("  Answer: n + (n-1) + (n-2) + ... + 1 = n(n+1)/2 = O(n^2)")
print("  (Still quadratic even though inner loop shrinks!)")
print()
print("Function 3: Halving n each iteration")
print("  Answer: O(log n) -- dividing by 2 each step")
print()
print("Function 4: Divide in half recursively (like merge sort)")
print("  Answer: O(n log n) -- log n levels, O(n) work per level")
````

# Section 4: Space Complexity

## What is Space Complexity?

Space complexity measures **how much extra memory** your algorithm uses
as input grows. It does NOT count the input itself.

| Pattern | Space | Why |
|---------|-------|-----|
| Fixed variables (int, pointer) | O(1) | Doesn't grow with input |
| New array of size n | O(n) | Grows linearly |
| 2D matrix of n x n | O(n^2) | Grows quadratically |
| Recursive call stack of depth n | O(n) | Each call uses stack frame |
| Hash map storing n items | O(n) | Grows linearly |

## Time-Space Trade-off

Often you can trade memory for speed:

| Approach | Time | Space | Example |
|----------|------|-------|---------|
| Two loops (no extra storage) | O(n^2) | O(1) | Naive duplicate check |
| Hash set (extra storage) | O(n) | O(n) | Set-based duplicate check |
| Sort in-place | O(n log n) | O(1)* | Sort then check neighbors |

*Note: Python's sort uses O(n) space internally (Timsort).

**Interview tip:** Mention the trade-off. Interviewers love when you say:
"I can do O(n) time with O(n) space using a hash map, or O(n log n) time
with O(1) space by sorting first. Which would you prefer?" 

````python
# Space complexity examples
print("SPACE COMPLEXITY EXAMPLES")
print("=" * 60)

# O(1) space -- constant extra memory
print("\n1. O(1) space -- only fixed variables")
def find_max(arr):
    max_val = arr[0]  # Just one variable
    for num in arr:
        if num > max_val:
            max_val = num
    return max_val
print("   find_max: only uses 'max_val' -- O(1) space")

# O(n) space -- creates new collection proportional to input
print("\n2. O(n) space -- new collection of size n")
def double_all(arr):
    result = []  # New array grows to size n
    for x in arr:
        result.append(x * 2)
    return result
print("   double_all: creates new array of size n -- O(n) space")

# O(n) space -- hash map
print("\n3. O(n) space -- hash map stores up to n items")
def two_sum(nums, target):
    seen = {}  # Could store up to n entries
    for i, num in enumerate(nums):
        complement = target - num
        if complement in seen:
            return [seen[complement], i]
        seen[num] = i
    return []
print("   two_sum: hash map could hold n items -- O(n) space")

# O(n) space -- recursion call stack
print("\n4. O(n) space -- recursion depth")
def factorial(n):
    if n <= 1:
        return 1
    return n * factorial(n - 1)  # n stack frames!
print("   factorial(n): call stack goes n levels deep -- O(n) space")

# O(1) space -- in-place modification
print("\n5. O(1) space -- modify input in-place")
def reverse_in_place(arr):
    left, right = 0, len(arr) - 1
    while left < right:
        arr[left], arr[right] = arr[right], arr[left]
        left += 1
        right -= 1
print("   reverse_in_place: only uses two pointers -- O(1) space")
````

# Section 5: Amortized Analysis

## What is Amortized Analysis?

Some operations are USUALLY fast but OCCASIONALLY slow.
Amortized analysis averages the cost over many operations.

**Classic example: Python list.append()**

- Most appends: O(1) -- just add to the end
- Occasionally: O(n) -- when the internal array is full, Python:
  1. Allocates a new, larger array (typically 2x)
  2. Copies all elements to the new array
  3. Adds the new element

Even though some appends are O(n), the AVERAGE over n appends is O(1).
We say append is **amortized O(1)**.

## Intuition

Imagine you're saving money:
- 99 days: save $1 (cheap operation)
- 1 day: spend $100 on something big (expensive operation)

Total over 100 days: $99 saved - $100 spent? No -- the analogy is:
- You "bank" cheap operations until you can "afford" the expensive one
- Average cost per operation = total cost / number of operations = O(1)

````python
# Amortized analysis -- demonstrating list.append()
print("AMORTIZED ANALYSIS -- list.append()")
print("=" * 60)
print()

import sys

# Watch how list capacity grows
print("How Python list grows internally:")
print(f"  {'Elements':<12} {'Actual Size (bytes)':<22} {'Capacity grew?':<15}")
print(f"  {'-'*50}")

lst = []
prev_size = sys.getsizeof(lst)
for i in range(20):
    lst.append(i)
    curr_size = sys.getsizeof(lst)
    grew = "YES (realloc!)" if curr_size > prev_size else ""
    if grew or i < 5:
        print(f"  {i+1:<12} {curr_size:<22} {grew:<15}")
    prev_size = curr_size

print()
print("  Key insight: Size jumps happen rarely (exponential growth).")
print("  Each jump copies all elements (O(n)), but happens every ~n/2 appends.")
print("  Amortized cost per append = O(1).")

# Timing demonstration
print()
print("Timing 1M appends (should be ~linear total, O(1) each):")
sizes = [100_000, 500_000, 1_000_000, 5_000_000]
for size in sizes:
    start = time.perf_counter()
    lst = []
    for i in range(size):
        lst.append(i)
    elapsed = time.perf_counter() - start
    per_op = elapsed / size * 1_000_000  # microseconds
    print(f"  {size:>10,} appends: {elapsed:.4f}s ({per_op:.3f} us/append)")
````

# Section 6: Comparing Algorithms -- Which is Better?

## Decision Framework

1. **Compare time complexity first** -- O(n) beats O(n^2) for large n
2. **If same time, compare space** -- O(1) space is better than O(n)
3. **Consider constants for small n** -- O(n^2) with tiny constant can beat O(n log n) for n < 50
4. **Consider the problem constraints:**
   - n <= 100: O(n^2) is fine, even O(n^3) works
   - n <= 10,000: O(n^2) is borderline
   - n <= 1,000,000: need O(n log n) or O(n)
   - n <= 10^9: need O(n) or O(log n)

## Quick Reference for Interviews

| Constraint | Acceptable Complexity |
|-----------|----------------------|
| n <= 10 | O(n!), O(2^n) |
| n <= 20 | O(2^n) |
| n <= 100 | O(n^3) |
| n <= 1,000 | O(n^2) |
| n <= 100,000 | O(n log n) |
| n <= 10,000,000 | O(n) |
| n > 10,000,000 | O(log n), O(1) |

````python
# Head-to-head comparison: O(n^2) vs O(n log n) vs O(n)
print("HEAD-TO-HEAD: Finding duplicates three ways")
print("=" * 60)
print()

import random, time

def has_dup_n2(arr):
    """O(n^2) -- compare every pair."""
    for i in range(len(arr)):
        for j in range(i+1, len(arr)):
            if arr[i] == arr[j]:
                return True
    return False

def has_dup_sort(arr):
    """O(n log n) -- sort then check neighbors."""
    s = sorted(arr)
    for i in range(len(s)-1):
        if s[i] == s[i+1]:
            return True
    return False

def has_dup_set(arr):
    """O(n) -- use a hash set."""
    return len(arr) != len(set(arr))

print(f"  {'Size':<10} {'O(n^2)':<15} {'O(n log n)':<15} {'O(n)':<15}")
print(f"  {'-'*55}")

for size in [1000, 5000, 10000, 50000]:
    arr = list(range(size))  # No dups -- worst case

    if size <= 10000:
        start = time.perf_counter()
        has_dup_n2(arr)
        t1 = time.perf_counter() - start
        t1_str = f"{t1:.4f}s"
    else:
        t1_str = "too slow"

    start = time.perf_counter()
    has_dup_sort(arr)
    t2 = time.perf_counter() - start

    start = time.perf_counter()
    has_dup_set(arr)
    t3 = time.perf_counter() - start

    print(f"  {size:<10,} {t1_str:<15} {t2:.4f}s{'':<9} {t3:.4f}s")

print()
print("  At n=50,000: O(n^2) would take ~40 seconds.")
print("  O(n log n) takes fractions of a second.")
print("  O(n) is even faster.")
print("  This is the power of choosing the right algorithm.")
````

# Section 7: Complexity in Data Engineering

## Why This Matters for Your Day Job

| Scenario | Bad Choice | Good Choice |
|----------|-----------|-------------|
| Dedup 100M records | Nested loop O(n^2) | Hash set O(n) |
| Sort 50M rows | Bubble sort O(n^2) | Merge/Tim sort O(n log n) |
| Find a record by ID | Linear scan O(n) | Hash index O(1) |
| Join two tables (1M x 10M) | Nested loop O(n*m) | Hash join O(n + m) |
| Partition pruning | Full scan O(n) | Binary search O(log n) |

## Spark Partitioning = Divide and Conquer

When Spark processes data across 200 partitions:
- Total data: 1 billion rows
- Per partition: 5 million rows
- Algorithm per partition: O(n log n) sort

Without partitioning: O(n log n) where n = 1 billion = slow
With partitioning: 200 * O(n/200 * log(n/200)) = much faster in parallel

This is why understanding complexity helps you tune Spark jobs!

````python
# DE example: Why hash joins are O(n+m) vs nested loop O(n*m)
print("DATA ENGINEERING: Hash Join vs Nested Loop Join")
print("=" * 60)
print()

# Simulate two tables
import random
orders = [(i, random.randint(1, 1000)) for i in range(10000)]   # 10K orders
customers = [(i, f"Customer_{i}") for i in range(1000)]         # 1K customers

# Nested loop join: O(n * m)
def nested_loop_join(orders, customers):
    results = []
    for order_id, cust_id in orders:
        for c_id, name in customers:
            if cust_id == c_id:
                results.append((order_id, name))
                break
    return results

# Hash join: O(n + m)
def hash_join(orders, customers):
    # Build phase: O(m) -- create hash map from smaller table
    cust_map = {c_id: name for c_id, name in customers}
    # Probe phase: O(n) -- look up each order
    results = []
    for order_id, cust_id in orders:
        if cust_id in cust_map:
            results.append((order_id, cust_map[cust_id]))
    return results

# Compare
start = time.perf_counter()
r1 = nested_loop_join(orders, customers)
t1 = time.perf_counter() - start

start = time.perf_counter()
r2 = hash_join(orders, customers)
t2 = time.perf_counter() - start

print(f"  Nested loop join (O(n*m)):  {t1:.4f}s  ({len(r1)} results)")
print(f"  Hash join (O(n+m)):         {t2:.4f}s  ({len(r2)} results)")
print(f"  Speedup: {t1/t2:.0f}x faster")
print()
print("  This is EXACTLY why Spark defaults to hash joins!")
print("  And why broadcast joins are even faster (small table fits in memory).")
````

# Section 9: Summary & Cheat Sheet

## Big-O Cheat Sheet

```
FASTEST
  O(1)        -- dict lookup, array index
  O(log n)    -- binary search
  O(n)        -- single loop, hash operations
  O(n log n)  -- sorting (best possible for comparison sort)
  O(n^2)      -- nested loops (avoid for large n)
  O(2^n)      -- recursion without memoization
  O(n!)       -- generating all permutations
SLOWEST
```

## Rules for Analysis

1. **Drop constants:** O(3n) = O(n)
2. **Drop lower terms:** O(n^2 + n) = O(n^2)
3. **Multiply nested operations:** Loop in loop = O(n^2)
4. **Add sequential operations:** Loop then loop = O(n) + O(n) = O(n)
5. **Recursion:** T(n) = 2*T(n/2) + O(n) = O(n log n) (divide & conquer)

## Space Complexity Quick Guide

- Variables (int, pointer): O(1)
- Array/list of size n: O(n)
- Hash map with n keys: O(n)
- Recursion depth d: O(d) for call stack
- 2D matrix n x m: O(n * m)

## What to Say in Interviews

> "This solution is O(n) time and O(n) space because we iterate once
> through the array and store at most n elements in the hash map.
> We could reduce space to O(1) by sorting first, which would make it
> O(n log n) time. Given the constraints (n up to 10^5), both are fine,
> but the hash map approach is faster."
