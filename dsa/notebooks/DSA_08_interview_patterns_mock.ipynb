  YES -> Backtracking

## Mock Problem 1: LRU Cache (Medium)

**Problem:** Design a data structure that supports:
- `get(key)`: Return value if key exists, else -1
- `put(key, value)`: Insert/update. If at capacity, evict least recently used.

Both operations must be O(1).

**Constraints:** capacity >= 1

**DE Connection:** This is exactly how Spark caches partitions and how
database buffer pools work!

````python
# ============================================
# MOCK 1: LRU Cache -- Try it yourself first! (25 min)
# ============================================
# Hint: Use a hash map + doubly linked list
# Or in Python: use OrderedDict

# Your solution here:

# ----- SOLUTION BELOW -----
from collections import OrderedDict

class LRUCache:
    """
    OrderedDict maintains insertion order.
    Move to end on access, pop from front on eviction.
    Time: O(1) for both get and put.
    """
    def __init__(self, capacity):
        self.cache = OrderedDict()
        self.capacity = capacity

    def get(self, key):
        if key not in self.cache:
            return -1
        self.cache.move_to_end(key)  # Mark as recently used
        return self.cache[key]

    def put(self, key, value):
        if key in self.cache:
            self.cache.move_to_end(key)
        self.cache[key] = value
        if len(self.cache) > self.capacity:
            self.cache.popitem(last=False)  # Evict oldest

# Test
lru = LRUCache(2)
lru.put(1, 1)
lru.put(2, 2)
print(f"Mock 1: LRU Cache")
print(f"  get(1) = {lru.get(1)}")      # 1
lru.put(3, 3)                          # Evicts key 2
print(f"  get(2) = {lru.get(2)}")      # -1 (evicted)
lru.put(4, 4)                          # Evicts key 1
print(f"  get(1) = {lru.get(1)}")      # -1 (evicted)
print(f"  get(3) = {lru.get(3)}")      # 3
print(f"  get(4) = {lru.get(4)}")      # 4
print("  Pattern: Hash Map + Ordered structure")
````

## Mock Problem 2: Serialize and Deserialize Binary Tree (Hard)

**Problem:** Design a way to convert a binary tree to a string
and back to the original tree structure.

**Constraints:** The tree can have any values including negatives.

**DE Connection:** This is serialization -- how data is stored in formats
like JSON, Parquet, or transmitted over the wire.

````python
# ============================================
# MOCK 2: Serialize/Deserialize Binary Tree (25 min)
# ============================================
from collections import deque

class TreeNode:
    def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right

def serialize(root):
    """Encode tree to string using BFS (level-order)."""
    if not root:
        return ""
    result = []
    queue = deque([root])
    while queue:
        node = queue.popleft()
        if node:
            result.append(str(node.val))
            queue.append(node.left)
            queue.append(node.right)
        else:
            result.append("null")
    return ",".join(result)

def deserialize(data):
    """Decode string back to tree."""
    if not data:
        return None
    values = data.split(",")
    root = TreeNode(int(values[0]))
    queue = deque([root])
    i = 1
    while queue and i < len(values):
        node = queue.popleft()
        if values[i] != "null":
            node.left = TreeNode(int(values[i]))
            queue.append(node.left)
        i += 1
        if i < len(values) and values[i] != "null":
            node.right = TreeNode(int(values[i]))
            queue.append(node.right)
        i += 1
    return root

# Test
root = TreeNode(1, TreeNode(2), TreeNode(3, TreeNode(4), TreeNode(5)))
encoded = serialize(root)
decoded = deserialize(encoded)
print(f"Mock 2: Serialize/Deserialize Tree")
print(f"  Serialized: {encoded}")
print(f"  Deserialized root: {decoded.val}")
print(f"  Left: {decoded.left.val}, Right: {decoded.right.val}")
print("  Pattern: BFS traversal + string encoding")
````

## Mock Problem 3: Word Search in Grid (Medium)

**Problem:** Given an m x n grid of characters and a word, find if the word
exists in the grid. The word can be constructed from sequentially adjacent
cells (horizontal or vertical). Same cell cannot be used twice.

**Example:**
```
board = [["A","B","C","E"],
         ["S","F","C","S"],
         ["A","D","E","E"]]
word = "ABCCED" -> True
```

````python
# ============================================
# MOCK 3: Word Search (25 min)
# ============================================

def exist(board, word):
    """
    DFS/Backtracking from each cell.
    Time: O(m * n * 4^L) where L = word length
    Space: O(L) for recursion
    """
    rows, cols = len(board), len(board[0])

    def dfs(r, c, idx):
        if idx == len(word):
            return True
        if (r < 0 or r >= rows or c < 0 or c >= cols or
            board[r][c] != word[idx]):
            return False

        # Mark as visited
        temp = board[r][c]
        board[r][c] = "#"

        # Explore 4 directions
        found = (dfs(r+1, c, idx+1) or dfs(r-1, c, idx+1) or
                 dfs(r, c+1, idx+1) or dfs(r, c-1, idx+1))

        # Backtrack
        board[r][c] = temp
        return found

    for r in range(rows):
        for c in range(cols):
            if dfs(r, c, 0):
                return True
    return False

board = [["A","B","C","E"],["S","F","C","S"],["A","D","E","E"]]
print(f"Mock 3: Word Search")
print(f"  Search 'ABCCED': {exist([row[:] for row in board], 'ABCCED')}")
print(f"  Search 'SEE': {exist([row[:] for row in board], 'SEE')}")
print(f"  Search 'ABCB': {exist([row[:] for row in board], 'ABCB')}")
print("  Pattern: Backtracking (DFS on grid)")
````

## Mock Problem 4: Design a Rate Limiter (Medium - DE Focused)

**Problem:** Implement a rate limiter that allows at most N requests
per second per user.

- `is_allowed(user_id, timestamp)`: Returns True if request is allowed.

**DE Connection:** Rate limiting is essential in:
- API ingestion pipelines
- Kafka producer throttling
- Spark structured streaming rate control

````python
# ============================================
# MOCK 4: Rate Limiter (25 min)
# ============================================
from collections import defaultdict, deque

class RateLimiter:
    """
    Sliding window rate limiter.
    Allows max_requests per window_seconds per user.
    Time: O(1) amortized, Space: O(users * max_requests)
    """
    def __init__(self, max_requests, window_seconds):
        self.max_requests = max_requests
        self.window = window_seconds
        self.requests = defaultdict(deque)  # user -> deque of timestamps

    def is_allowed(self, user_id, timestamp):
        # Remove expired timestamps
        user_queue = self.requests[user_id]
        while user_queue and timestamp - user_queue[0] >= self.window:
            user_queue.popleft()

        if len(user_queue) < self.max_requests:
            user_queue.append(timestamp)
            return True
        return False

# Test: 3 requests per 1 second
limiter = RateLimiter(max_requests=3, window_seconds=1.0)
print("Mock 4: Rate Limiter (3 req/sec)")

tests = [
    ("user1", 0.0, True),
    ("user1", 0.3, True),
    ("user1", 0.6, True),
    ("user1", 0.9, False),  # 4th in same second -- blocked!
    ("user2", 0.9, True),   # Different user -- allowed
    ("user1", 1.1, True),   # First request expired, window shifted
]
for user, ts, expected in tests:
    result = limiter.is_allowed(user, ts)
    status = "PASS" if result == expected else "FAIL"
    print(f"  [{status}] user={user}, t={ts}: {'allowed' if result else 'BLOCKED'}")

print("  Pattern: Queue (sliding window of timestamps)")
````

## Mock Problem 5: Pipeline Dependency Resolver (Medium - DE Focused)

**Problem:** Given a list of data pipeline tasks with dependencies,
return a valid execution order. If circular dependencies exist, return empty.

This is a direct mapping to what Airflow/Databricks Workflows does internally.

**Example:**
```
tasks = ["extract", "transform", "load", "validate", "report"]
deps = [("transform", "extract"), ("load", "transform"),
        ("validate", "load"), ("report", "validate")]
Output: ["extract", "transform", "load", "validate", "report"]
```

````python
# ============================================
# MOCK 5: Pipeline Dependency Resolver (25 min)
# ============================================
from collections import defaultdict, deque

def resolve_pipeline(tasks, dependencies):
    """
    Topological sort with cycle detection (Kahn's algorithm).
    Time: O(V + E), Space: O(V + E)
    """
    graph = defaultdict(list)
    in_degree = {task: 0 for task in tasks}

    for task, dep in dependencies:
        graph[dep].append(task)
        in_degree[task] += 1

    # Start with tasks that have no dependencies
    queue = deque([t for t in tasks if in_degree[t] == 0])
    order = []

    while queue:
        task = queue.popleft()
        order.append(task)
        for next_task in graph[task]:
            in_degree[next_task] -= 1
            if in_degree[next_task] == 0:
                queue.append(next_task)

    if len(order) == len(tasks):
        return order
    return []  # Circular dependency!

# Test 1: Valid pipeline
tasks = ["extract", "transform", "load", "validate", "report"]
deps = [("transform", "extract"), ("load", "transform"),
        ("validate", "load"), ("report", "validate")]
print("Mock 5: Pipeline Dependency Resolver")
print(f"  Tasks: {tasks}")
print(f"  Order: {resolve_pipeline(tasks, deps)}")

# Test 2: Circular dependency
tasks2 = ["A", "B", "C"]
deps2 = [("A", "B"), ("B", "C"), ("C", "A")]
print(f"\n  Circular deps: {deps2}")
print(f"  Order: {resolve_pipeline(tasks2, deps2)} (empty = cycle!)")
print("  Pattern: Topological Sort (Kahn's algorithm)")
````

## Mock Problem 6: Deduplicate Data Stream (Medium - DE Focused)

**Problem:** Design a system that deduplicates records from a streaming
data source within a time window. Records have an ID and timestamp.

- `process(record_id, timestamp)`: Return True if this is the first time
  seeing this ID within the last `window` seconds.

**DE Connection:** This is exactly what Spark Structured Streaming's
`dropDuplicatesWithinWatermark()` does.

````python
# ============================================
# MOCK 6: Streaming Deduplication (25 min)
# ============================================

class StreamDeduplicator:
    """
    Deduplicates records within a time window.
    Uses a dict of {record_id: last_seen_timestamp} + cleanup.
    Time: O(1) amortized, Space: O(unique records in window)
    """
    def __init__(self, window_seconds):
        self.window = window_seconds
        self.seen = {}  # record_id -> timestamp
        self.cleanup_interval = window_seconds * 2
        self.last_cleanup = 0

    def process(self, record_id, timestamp):
        # Periodic cleanup of expired entries
        if timestamp - self.last_cleanup > self.cleanup_interval:
            self._cleanup(timestamp)
            self.last_cleanup = timestamp

        # Check if duplicate within window
        if record_id in self.seen:
            if timestamp - self.seen[record_id] < self.window:
                return False  # Duplicate!
        
        self.seen[record_id] = timestamp
        return True  # First occurrence (or re-appeared after window)

    def _cleanup(self, current_time):
        expired = [rid for rid, ts in self.seen.items()
                   if current_time - ts >= self.window]
        for rid in expired:
            del self.seen[rid]

# Test
dedup = StreamDeduplicator(window_seconds=5.0)
print("Mock 6: Stream Deduplication (5 sec window)")

events = [
    ("order_1", 1.0, True),   # First time
    ("order_2", 2.0, True),   # First time
    ("order_1", 3.0, False),  # Duplicate (within 5 sec)
    ("order_3", 4.0, True),   # First time
    ("order_1", 7.0, True),   # order_1 reappears after 6 sec -- allowed!
]
for rid, ts, expected in events:
    result = dedup.process(rid, ts)
    status = "PASS" if result == expected else "FAIL"
    print(f"  [{status}] process('{rid}', t={ts}): {'NEW' if result else 'DUPLICATE'}")
print("  Pattern: Hash Map + Time-based expiration")
````

# Solutions for Mock Problems 7-10

````python
# SOLUTION 7: Minimum Window Substring
from collections import Counter

def min_window(s, t):
    """Sliding window with character frequency tracking. O(n)."""
    if not s or not t:
        return ""
    need = Counter(t)
    missing = len(t)
    left = 0
    best = ""

    for right, char in enumerate(s):
        if need[char] > 0:
            missing -= 1
        need[char] -= 1

        while missing == 0:
            window = s[left:right + 1]
            if not best or len(window) < len(best):
                best = window
            need[s[left]] += 1
            if need[s[left]] > 0:
                missing += 1
            left += 1

    return best

print("SOLUTION 7: Minimum Window Substring")
print(f"  min_window('ADOBECODEBANC', 'ABC') = '{min_window('ADOBECODEBANC', 'ABC')}'")
````

````python
# SOLUTION 8: Trie
class TrieNode:
    def __init__(self):
        self.children = {}
        self.is_end = False

class Trie:
    def __init__(self):
        self.root = TrieNode()

    def insert(self, word):
        node = self.root
        for char in word:
            if char not in node.children:
                node.children[char] = TrieNode()
            node = node.children[char]
        node.is_end = True

    def search(self, word):
        node = self._find(word)
        return node is not None and node.is_end

    def starts_with(self, prefix):
        return self._find(prefix) is not None

    def _find(self, prefix):
        node = self.root
        for char in prefix:
            if char not in node.children:
                return None
            node = node.children[char]
        return node

print("\nSOLUTION 8: Trie")
trie = Trie()
trie.insert("apple")
trie.insert("app")
print(f"  search('apple') = {trie.search('apple')}")
print(f"  search('app') = {trie.search('app')}")
print(f"  startsWith('app') = {trie.starts_with('app')}")
print(f"  search('banana') = {trie.search('banana')}")
````

````python
# SOLUTION 9: Maximum Profit Job Scheduling
import bisect

def job_scheduling(start, end, profit):
    """DP + Binary Search. Sort by end time, use bisect for previous job."""
    jobs = sorted(zip(end, start, profit))
    n = len(jobs)
    dp = [0] * (n + 1)  # dp[i] = max profit considering first i jobs

    for i in range(1, n + 1):
        e, s, p = jobs[i-1]
        # Find last job that ends before this one starts
        ends = [jobs[j][0] for j in range(i-1)]
        j = bisect.bisect_right(ends, s)
        dp[i] = max(dp[i-1], dp[j] + p)

    return dp[n]

print("\nSOLUTION 9: Job Scheduling")
starts = [1, 2, 3, 3]
ends = [3, 4, 5, 6]
profits = [50, 10, 40, 70]
print(f"  Max profit: {job_scheduling(starts, ends, profits)}")
````

````python
# SOLUTION 10: TTL Key-Value Store
import time as time_module

class TTLKeyValueStore:
    def __init__(self):
        self.store = {}  # key -> (value, expiry_time or None)

    def set(self, key, value, ttl=None, current_time=0):
        expiry = current_time + ttl if ttl else None
        self.store[key] = (value, expiry)

    def get(self, key, current_time=0):
        if key not in self.store:
            return None
        value, expiry = self.store[key]
        if expiry is not None and current_time >= expiry:
            del self.store[key]
            return None
        return value

print("\nSOLUTION 10: TTL Key-Value Store")
kv = TTLKeyValueStore()
kv.set("api_key", "abc123", ttl=10, current_time=0)
kv.set("config", "prod", current_time=0)  # No TTL

print(f"  get('api_key', t=5) = {kv.get('api_key', current_time=5)}")
print(f"  get('api_key', t=15) = {kv.get('api_key', current_time=15)}")  # Expired
print(f"  get('config', t=100) = {kv.get('config', current_time=100)}")  # No TTL
print(f"  get('missing', t=0) = {kv.get('missing', current_time=0)}")
````
