# Section 1: The Data Engineering Career Ladder

## Levels & Expectations

| Level | Title | YOE* | Scope | Key Differentiator |
|-------|-------|------|-------|-------------------|
| L3 | Junior DE | 0-2 | Single tasks, guided | Writes working code with help |
| L4 | Data Engineer | 2-4 | Features, own pipeline | Owns end-to-end pipelines independently |
| L5 | Senior DE | 4-7 | Projects, mentoring | Designs systems, makes tech decisions |
| L6 | Staff DE | 7-12 | Organization-wide | Sets standards, influences architecture |
| L7 | Principal DE | 12+ | Company-wide strategy | Industry thought leader, defines direction |

*YOE = typical years of experience (varies significantly by individual)

## What Each Level Looks Like in Practice

```
JUNIOR (L3):
  - Writes Spark jobs from specs given by senior
  - Fixes bugs in existing pipelines
  - Adds columns to existing tables
  - Asks questions daily, gets code reviewed

MID (L4):
  - Owns 2-3 pipelines end-to-end (build, deploy, monitor)
  - Designs schema for new data sources
  - Handles incidents independently
  - Proposes improvements, doesn't just execute

SENIOR (L5):
  - Designs new pipeline architectures from business requirements
  - Mentors juniors, reviews their code
  - Makes technology choices (Kafka vs Kinesis, dbt vs Spark)
  - Identifies and solves problems nobody assigned
  - Influences team roadmap
  - Leads projects spanning multiple sprints

STAFF (L6):
  - Defines the data platform architecture for the org
  - Sets coding standards, review practices, testing requirements
  - Solves cross-team technical challenges
  - Evaluates and introduces new technologies
  - Technical partner to engineering directors
  - Drives 6-12 month technical strategy

PRINCIPAL (L7):
  - Sets multi-year technical vision
  - Influences industry (conference talks, open-source)
  - Reports to VP/CTO level
  - Defines what "good" looks like across the company
  - Rarely writes code, always designs systems
```

## Skills Matrix by Level

| Skill | Junior | Mid | Senior | Staff |
|-------|--------|-----|--------|-------|
| SQL | Basic queries | Complex (windows, CTEs) | Optimization, design | Teaches others |
| Python | Scripts | Production code | Architecture | Standards |
| Spark | Use existing | Build from scratch | Tune & optimize | Design frameworks |
| System Design | N/A | Understand | Design & present | Define patterns |
| Communication | Team updates | Stakeholder docs | Present to leadership | External thought leader |
| Mentoring | N/A | Help peers | Mentor juniors | Mentor seniors |
| Scope | Task | Feature/pipeline | Project/system | Organization |

# Section 3: The DE Interview Process

## Typical Interview Loop (5 Rounds)

```
PHONE SCREEN (30-45 min):
  - Recruiter: background, motivations, salary range
  - Optional: quick SQL or Python screen (easy-medium)

TECHNICAL SCREEN (45-60 min):
  - Live coding: SQL (2 problems) or Python (1-2 problems)
  - Level: Easy-Medium LeetCode equivalent
  - Focus: Can you write working code under pressure?

SYSTEM DESIGN (45-60 min):
  - "Design a data pipeline for X"
  - Level: Matches your seniority
  - Focus: Trade-offs, scalability, communication

CODING DEEP DIVE (45-60 min):
  - Harder coding: Spark optimization, complex SQL
  - OR: Take-home review + extensions
  - Focus: Depth of knowledge, production thinking

BEHAVIORAL (45-60 min):
  - "Tell me about a time when..."
  - Focus: Leadership, conflict resolution, failure handling
  - Usually with hiring manager

OPTIONAL: Hiring Manager Chat (30 min)
  - Role expectations, team structure, growth opportunities
  - Your chance to evaluate THEM
```

## What's Tested at Each Level

| Round | Junior (L3-L4) | Senior (L5) | Staff (L6) |
|-------|---------------|-------------|------------|
| SQL | Medium (joins, windows) | Hard (optimization, cohort) | Architecture review |
| Python | Medium (data structures) | Hard (design patterns) | System design in Python |
| System Design | Explain existing | Design from scratch | Define standards |
| Behavioral | Teamwork, learning | Technical leadership | Org-wide influence |
| Spark | Basic transforms | Performance tuning | Architecture decisions |

# Section 4: Behavioral Interviews (STAR Method)

## The STAR Framework

```
S - SITUATION: Set the context (1-2 sentences)
    "We had a daily ETL pipeline that was consistently failing..."

T - TASK: What was YOUR responsibility?
    "As the senior DE, I was responsible for diagnosing and fixing it..."

A - ACTION: What specifically did YOU do? (most important!)
    "I analyzed the Spark UI, found a data skew issue, implemented salting..."

R - RESULT: Quantified outcome
    "Pipeline runtime dropped from 4 hours to 25 minutes. Zero failures
     in the 6 months since. Other teams adopted the same pattern."
```

## Top 10 Behavioral Questions for DE

| # | Question | What They're Testing |
|---|----------|---------------------|
| 1 | Tell me about a pipeline failure you debugged | Problem-solving, composure |
| 2 | How did you handle a disagreement on technical approach? | Collaboration, persuasion |
| 3 | Describe a time you improved a system's performance significantly | Technical depth, impact |
| 4 | Tell me about a project that failed. What did you learn? | Self-awareness, growth |
| 5 | How do you handle conflicting priorities from stakeholders? | Prioritization, communication |
| 6 | Describe a time you mentored someone | Leadership, patience |
| 7 | Tell me about the most complex pipeline you've built | Technical scope, architecture |
| 8 | How do you ensure data quality in your pipelines? | Production mindset |
| 9 | Describe a time you had to learn a new technology quickly | Adaptability, self-learning |
| 10 | How do you handle working with messy/undocumented data? | Pragmatism, problem-solving |

## Example STAR Answer

```
Q: "Tell me about a time you significantly improved pipeline performance."

S: "Our daily orders pipeline was taking 6 hours to run on a 30-node EMR
    cluster, frequently missing our 6 AM SLA for the finance team's reports."

T: "As the senior DE, I was tasked with getting it under 1 hour without
    increasing our cloud spend."

A: "I took a three-pronged approach:
    1. Analyzed Spark UI: found 90% of time was in one shuffle stage with
       extreme data skew (one customer had 10M orders vs average 100).
    2. Implemented salting technique for the skewed join key, distributing
       the hot key across 50 partitions.
    3. Converted from Parquet to Delta Lake with Liquid Clustering on
       customer_id, eliminating the need for repartitioning.
    4. Added predicate pushdown by restructuring filters before joins.
    
    I also set up Spark UI monitoring alerts so we'd catch regressions early."

R: "Pipeline time dropped from 6 hours to 35 minutes — a 10x improvement.
    We actually REDUCED the cluster from 30 to 12 nodes (60% cost saving of
    $4K/month). The pattern was documented and adopted by 3 other teams.
    We haven't missed SLA since (6 months and counting)."
```

## DE Compensation Ranges (US, 2025-2026)

| Level | Base Salary | Total Comp (incl. equity/bonus) | Where |
|-------|-------------|------|-------|
| Junior (L3) | $90K - $130K | $100K - $150K | Mid-market |
| Mid (L4) | $130K - $170K | $150K - $220K | Mid-market |
| Senior (L5) | $160K - $220K | $200K - $350K | FAANG/top-tier |
| Staff (L6) | $200K - $280K | $300K - $500K | FAANG/top-tier |
| Principal (L7) | $250K+ | $500K - $800K+ | FAANG only |

*Varies significantly by location, company stage, and market conditions*

## Negotiation Framework

```
1. NEVER give a number first.
   "I'm focused on finding the right fit. What's the range for this role?"
   
2. When you get an offer, DON'T accept immediately.
   "Thank you! I'm very excited. I'd like a day to review the full package."

3. Research using:
   - levels.fyi (best for tech companies)
   - Glassdoor, Blind, Teamblind
   - LinkedIn salary insights
   - Recruiter conversations (ask range in first call)

4. Counter with data:
   "Based on my research and the scope of this role, I was expecting
    [X]. Given my experience with [specific relevant skill], and the
    impact I plan to make on [specific project they mentioned], would
    you be able to get closer to [target number]?"

5. Negotiate the FULL package:
   - Base salary (most important for long-term)
   - Signing bonus (one-time, easier for company to give)
   - Equity/RSUs (important at growth companies)
   - Annual bonus target
   - Remote flexibility
   - Start date (more time = more prep = better start)
   - Title (affects future negotiations)
   - Education budget / conference attendance

6. Get it in writing before accepting verbally.
```

# Section 7: Job Search Strategy

## The Job Search Funnel

```
APPLICATIONS (100)
    ↓ 20% response rate
RECRUITER SCREENS (20)
    ↓ 50% pass rate
TECHNICAL SCREENS (10)
    ↓ 50% pass rate
ON-SITE LOOPS (5)
    ↓ 33% offer rate
OFFERS (1-2)
    ↓ negotiate
ACCEPTED (1)

Timeline: 4-12 weeks from start to offer
Key: VOLUME matters. Apply to 5-10 roles/week minimum.
```

## Where to Find DE Jobs

| Source | Effectiveness | Notes |
|--------|--------------|-------|
| Referrals | Highest (50%+ interview rate) | Network actively |
| LinkedIn (direct apply) | Medium | Optimize profile for recruiters |
| Company career pages | Medium | Apply directly (skip ATS filters) |
| AngelList / Wellfound | Good for startups | Higher equity, lower base |
| levels.fyi job board | Good for top companies | Comp-transparent |
| Recruiter outreach | Passive (let them come) | Keep LinkedIn updated |
| Blind/TeamBlind | Good for referrals | Ask for referrals directly |

## 30-60-90 Day Plan (Your First 3 Months)

```
FIRST 30 DAYS: Learn & Listen
  - Understand the data architecture (draw diagrams!)
  - Meet every team member 1:1
  - Read existing pipeline code (don't judge yet)
  - Complete onboarding tasks, get dev environment running
  - Identify one small win (fix a bug, improve a doc)
  - Ask: "What's the biggest pain point right now?"

DAYS 31-60: Contribute & Propose
  - Own a pipeline end-to-end (fix, improve, monitor)
  - Propose one improvement (not a rewrite!)
  - Start contributing to code reviews
  - Document something that was undocumented
  - Build relationships with stakeholders

DAYS 61-90: Drive & Lead
  - Lead a small project or improvement initiative
  - Present your work to the team (lunch & learn)
  - Identify your first "big win" for the next quarter
  - Give your manager concrete goals for next review period
  - Mentor a newer team member (if applicable)
```

## Networking (The Underrated Superpower)

```
SHORT-TERM: Get referrals for current job search
  - LinkedIn: connect with DEs at target companies
  - Message: "I'm exploring DE roles. Would you be open to a 15-min chat
    about what it's like working on the data team at [Company]?"
  - After chat: "Would you be comfortable referring me? Here's my resume."
  - Referral = 50%+ chance of interview (vs 5% cold apply)

LONG-TERM: Build your brand
  - Write technical blog posts (1-2/month)
  - Contribute to open source
  - Share learnings on LinkedIn/Twitter
  - Attend/speak at meetups (local or virtual)
  - Answer questions on Stack Overflow / Reddit /r/dataengineering
```
